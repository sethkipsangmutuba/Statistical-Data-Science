## **3. The Logic of Hypothesis Testing**

We test whether observed data provide enough evidence to reject a null claim about a population parameter.

### **Four Steps in Hypothesis Testing**

A. State the null hypothesis ($H_0$) and alternative hypothesis ($H_1$)  
B. Compute the test statistic  
C. Determine the p-value and interpret it  
D. Compare with significance level ($\alpha$) and make a decision

---



## **4. Worked Example**

### **Problem Setup**
A lecturer claims that the population mean exam score is $\mu_0 = 75$.  
We take a random sample of $n = 50$ students and obtain a sample mean of $\bar{x} = 73$.  
The population standard deviation is known to be $\sigma = 10$.  
We test this claim at a significance level $\alpha = 0.05$, using a two-sided z-test.

---

## **5. Step A: Hypotheses**

We test:  

**$H_0$: $\mu = \mu_0 = 75$ (no difference)**  
**$H_1$: $\mu \neq 75$ (mean differs from 75)**

---



## **6. Step B: Compute the Test Statistic**

The test statistic standardizes the distance between the sample mean and the hypothesized mean:

$$
z = \frac{\bar{X} - \mu_0}{SE_{\bar{X}}}
$$

where the standard error of the sample mean is:

$$
SE_{\bar{X}} = \frac{\sigma}{\sqrt{n}}
$$

**Substitute values:**

$$
SE_{\bar{X}} = \frac{10}{\sqrt{50}} = \frac{10}{7.0711} = 1.4142
$$

Now compute:

$$
z = \frac{73 - 75}{1.4142} = -1.4142
$$

The z-value of -1.4142 indicates the sample mean is 1.41 standard errors below μ₀.

---

## **7. Step C: p-Value and Interpretation**

---

#### Conceptual Meaning

The **p-value** represents the probability, **assuming the null hypothesis ($H_0$)** is true, of observing a test statistic as extreme or more extreme than the one actually obtained from the sample.

In simpler terms:  
If $H_0$ were true, how likely is it to get a sample mean this far (or farther) from 75 just by random chance?

---

#### Formula for a Two-Sided Test

$$
p = 2P(Z \le -|z_{obs}|)
$$

The factor of **2** is because we’re considering both tails — deviations **above and below** the mean.

$|z_{obs}|$ means we only care about the **magnitude** of the z-value, not its direction (since it’s a two-sided test).

---

#### Substituting the Observed Value

$$
p = 2P(Z \le -1.4142)
$$

Now, from the **Standard Normal Table** (or software):

$$
P(Z \le -1.4142) = 0.07865
$$

This is the area (probability) to the left of $z = -1.4142$ on the standard normal curve.

---

#### Compute the Full Two-Tailed Probability

$$
p = 2(0.07865) = 0.1573
$$

**Interpretation:**  
There’s a **15.73% probability** of observing a sample mean this far (or farther) from 75 if the true mean were 75.

---

The observed difference between the sample mean (73) and the hypothesized mean (75) is **not statistically significant** at the 5% level.  
This means the data do **not provide sufficient evidence** to claim that the population mean exam score differs from 75.

---

#####In Plain Language:

> “The p-value is 0.1573, which is greater than 0.05. Therefore, we do not have enough statistical evidence to reject the null hypothesis.  
> The sample’s deviation from the hypothesized mean could reasonably be due to random variation.”

---
There is insufficient evidence, at the 5% significance level, to conclude that the population mean exam score differs from 75.

---



## **8. Step D: Power Analysis**

Now we ask:  
If the true mean were actually $\mu_1 = 78$, how likely would this test detect it?

This likelihood is called the **power** of the test, denoted $1 − \beta$.

## **Distribution Under the Alternative**
----

Under $H_1: \mu = \mu_1$, the sampling distribution of the test statistic becomes:

$$
Z^* = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}} \sim N(\delta, 1)
$$

where the non-centrality parameter $\delta$ is:

$$
\delta = \frac{\mu_1 - \mu_0}{\sigma / \sqrt{n}}
$$

**Compute:**

$$
\delta = \frac{78 - 75}{10 / \sqrt{50}} = \frac{3}{1.4142} = 2.1213
$$

---


---

## **Context**

We are working with a two-sided hypothesis test at a significance level of:

$$\alpha = 0.05$$

This means we are allowing a **5% risk** of making a **Type I error** — that is, rejecting a true null hypothesis.

From the standard normal distribution, the critical z-values that capture the middle 95% of the probability (leaving 2.5% in each tail) are:

$$z_{1 - \alpha/2} = 1.96$$

---

## **Decision Rule**

We reject $H_0$ when the test statistic lies beyond these critical values, that is, when:

$$|Z^*| > 1.96$$

This means:

- If the computed z-statistic is **less than -1.96** or **greater than +1.96**, we reject $H_0$.  
- Otherwise, we **fail to reject $H_0$**.

---

## **Concept of Power**

Power measures the probability that the test correctly rejects a false null hypothesis i.e., it detects a real effect when one truly exists.

Formally:

$$\text{Power} = P(\text{Reject } H_0 \mid H_1 \text{ is true})$$

For a two-sided z-test, this becomes:

$$\text{Power} = P(Z < -1.96 - \delta) + P(Z > 1.96 - \delta)$$

where:

- $Z$ follows a **standard normal distribution** under $H_1$ (i.e., when the true mean ≠ hypothesized mean)
- $\delta$ (delta) is the **non-centrality parameter**, representing the standardized shift between the true and hypothesized means:

$$\delta = \frac{(\mu_a - \mu_0)}{\sigma / \sqrt{n}}$$

---

## **Substitution**

We substitute the calculated $\delta$ (from our specific problem setup):

$$\text{Power} = P(Z < -4.0813) + P(Z > -0.1613)$$

Here:

- $-4.0813$ represents the **far left rejection boundary**,  
- $-0.1613$ represents the **shifted right boundary** under the true mean scenario.

---

## **Finding Probabilities from the Standard Normal Table**

Using standard normal ($Z$) tables:

$$P(Z < -4.0813) \approx 0.00002$$

and

$$P(Z > -0.1613) = 1 - P(Z \le -0.1613)$$

From the table:

$$P(Z \le -0.1613) = 0.4360$$

So:

$$P(Z > -0.1613) = 1 - 0.4360 = 0.5640$$

---

## **Combine the Two Tails**

$$\text{Power} = 0.00002 + 0.5640 = 0.564$$

---

## **Interpretation**

The test has a **power of approximately 0.564 (56.4%)** to detect a true population mean of 78 when the hypothesized mean is 75.

---

If the true mean is actually 78, the test will correctly reject the null hypothesis about **56% of the time**.

This also implies a **Type II error probability ($\beta$)** of:

$$\beta = 1 - \text{Power} = 1 - 0.564 = 0.436$$

That is, there’s about a **43.6% chance of failing to detect a true difference**  i.e., retaining $H_0$ when the mean really is 78.

---

With a sample size of **n = 50**, the test is only **moderately sensitive**.

There’s nearly a **44% chance of missing a real difference** if one exists, suggesting that a **larger sample** would improve the reliability and detection ability of this test.


---



## **9. Step E: Required Sample Size for 80% Power**

Researchers typically aim for $Power = 0.8$ (or $\beta = 0.2$).  

We use the formula for required sample size $n$ (two-sided z-test, known $\sigma$):

$$
n = \left(\frac{z_{1−\alpha/2} + z_{1−\beta}}{(\mu_1 − \mu_0)/\sigma}\right)^2
$$

**Given:**

$$
z_{1−\alpha/2} = 1.96, \quad z_{1−\beta} = 0.8416, \quad (\mu_1 − \mu_0)/\sigma = 0.3
$$

**Compute:**

$$
n = \left(\frac{1.96 + 0.8416}{0.3}\right)^2 = (9.3389)^2 = 87.21
$$

**Round up:**

$$
n ≈ 88
$$

**Interpretation:**  
To detect a 3-point difference (from 75 to 78) with 80% power at $\alpha = 0.05$, we require a sample of approximately **88 students**.

---

| Hypothesized mean | $\mu_0$ | 75 | Claimed population mean |
| Sample mean | $\bar{x}$ | 73 | Observed mean |

| Required $n$ for 0.8 power | Formula above | 88 | Minimum sample for strong test |

---

## **11. Conceptual Takeaways**

- **$\alpha$ (Significance level):** Probability of Type I error (rejecting a true $H_0$).  
- **$\beta$ (Type II error):** Probability of failing to reject a false $H_0$.  
- **Power ($1 − \beta$):** Probability of correctly rejecting $H_0$ when it’s false.  
- Increasing $n$ reduces both $\alpha$ and $\beta$, improving power.  
- Effect size $(\mu_1 − \mu_0)/\sigma$ strongly influences required sample size.  
- Always report: $n, \bar{x}, \sigma, z, p, \alpha, \text{decision}, \text{power}$.

---

Our analysis showed that while the evidence is insufficient to reject the lecturer’s claim ($p = 0.157 > 0.05$), the test’s low power ($\approx 56\%$) indicates that a larger sample—around $88$ students—would be needed to detect a true mean shift of $3$ points with high confidence.

#  Confusion Matrix with Statistical Mapping

| **Actual / Predicted** | **Predicted Positive** | **Predicted Negative** | **Row Total** |
|:------------------------|:-----------------------|:------------------------|:---------------------|
| **Actual Positive ($H_1$ true)** | **TP / Correct Detection / $1 - \beta$ (Power)** | **FN / Type II Error / $\beta$** | **TP + FN** |
| **Actual Negative ($H_0$ true)** | **FP / Type I Error / $\alpha$** | **TN / Correct Retention** | **FP + TN** |

#  Associated Machine Learning & Statistical Metrics

| **Metric** | **Formula** |
|:------------|:------------|
| **Accuracy** | $$\text{Accuracy} = \frac{TP + TN}{TP + FP + FN + TN}$$ |
| **Precision (Positive Predictive Value)** | $$\text{Precision} = \frac{TP}{TP + FP}$$ |
| **Recall (Sensitivity / True Positive Rate)** | $$\text{Recall} = \frac{TP}{TP + FN}$$ |
| **Specificity (True Negative Rate)** | $$\text{Specificity} = \frac{TN}{FP + TN}$$ |
| **F1 Score** | $$F_1 = 2 \times \frac{(\text{Precision} \times \text{Recall})}{(\text{Precision} + \text{Recall})}$$ |
| **Type I Error Rate ($\alpha$)** | $$\alpha = \frac{FP}{FP + TN}$$ |
| **Type II Error Rate ($\beta$)** | $$\beta = \frac{FN}{FN + TP}$$ |
| **Power ($1 - \beta$)** | $$1 - \beta = \frac{TP}{FN + TP}$$ |


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -------------------------------
# Given values
# -------------------------------
mu0 = 75
xbar = 73
sigma = 10
n = 50
alpha = 0.05

# -------------------------------
# Test Statistic
# -------------------------------
se = sigma / np.sqrt(n)
z_stat = (xbar - mu0) / se
p_value = 2 * norm.sf(abs(z_stat))
z_crit = norm.ppf(1 - alpha / 2)

decision = "Reject H₀" if p_value < alpha else "Fail to Reject H₀"

# -------------------------------
# Power Calculation
# -------------------------------
mu1 = 78
delta = (mu1 - mu0) / se
power = norm.sf(z_crit - delta) + norm.cdf(-z_crit - delta)

# Sample size for 80% power
target_power = 0.80
z_beta = norm.ppf(target_power)
effect = (mu1 - mu0) / sigma
n_required = ((norm.ppf(1 - alpha/2) + z_beta) / effect) ** 2
n_required = int(np.ceil(n_required))

# -------------------------------
# PRINT ALL METRICS
# -------------------------------
print("===== Hypothesis Test Summary =====")
print(f"Hypothesized Mean (μ₀): {mu0}")
print(f"Sample Mean (x̄): {xbar}")
print(f"Population SD (σ): {sigma}")
print(f"Sample Size (n): {n}")
print(f"Standard Error (SE): {se:.3f}")
print(f"z-Statistic: {z_stat:.4f}")
print(f"p-Value: {p_value:.4f}")
print(f"Decision (α = {alpha}): {decision}")
print("\n===== Power Analysis =====")
print(f"Assumed True Mean (μ₁): {mu1}")
print(f"Power (1 - β): {power:.3f}")
print(f"Required Sample Size for 80% Power: {n_required}")
print("===================================")
```

The observed mean is slightly below the hypothesized value, but the difference is not statistically significant at $\alpha = 0.05$ ($p = 0.157$). Thus, we fail to reject $H_0$. However, the test has low power ($\approx 56\%$) to detect a true mean of $78$, indicating that a larger sample ($\approx 88$) is needed.


```python
# -------------------------------
# VISUALIZATION
# -------------------------------
x = np.linspace(-4, 4, 400)
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# ---------- Plot 1 (H0) ----------
axs[0].plot(x, norm.pdf(x), color='black', linewidth=2, label="H₀ Distribution")
axs[0].fill_between(x, norm.pdf(x), where=(x > -z_crit) & (x < z_crit), color="lightblue", alpha=0.6, label="Fail to Reject H₀")
axs[0].fill_between(x, norm.pdf(x), where=(x <= -z_crit) | (x >= z_crit), color="lightcoral", alpha=0.6, label="Reject H₀")
axs[0].axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f"Observed z = {z_stat:.2f}")
axs[0].set_title("Sampling Distribution Under H₀")
axs[0].set_xlabel("z-value")
axs[0].set_ylabel("Density")
axs[0].legend()

# ---------- Plot 2 (H1 Shifted) ----------
axs[1].plot(x, norm.pdf(x - delta), color='green', linewidth=2, label="H₁ Distribution")
axs[1].axvline(z_crit, linestyle='--', color='red', linewidth=2, label="Critical Values")
axs[1].axvline(-z_crit, linestyle='--', color='red', linewidth=2)
axs[1].set_title("Sampling Distribution Under H₁")
axs[1].set_xlabel("z-value")
axs[1].set_ylabel("Density")
axs[1].legend()

# ---------- Plot 3 (Power) ----------
axs[2].plot(x, norm.pdf(x), color='blue', linewidth=2, label="H₀ Distribution")
axs[2].plot(x, norm.pdf(x - delta), color='green', linestyle='--', linewidth=2, label="H₁ Distribution")
axs[2].fill_between(x, norm.pdf(x - delta), where=(x <= -z_crit) | (x >= z_crit), color="lightgreen", alpha=0.6, label="Power (1 - β)")
axs[2].axvline(z_crit, linestyle=':', color='red', linewidth=2, label="Critical Values")
axs[2].axvline(-z_crit, linestyle=':', color='red', linewidth=2)
axs[2].set_title("Power Visualization")
axs[2].set_xlabel("z-value")
axs[2].set_ylabel("Density")
axs[2].legend()

plt.tight_layout()
plt.show()
```

# **Null and Alternative Hypotheses**

---

## **1. Introduction**

The first step in hypothesis testing is to translate the research question into two competing **statistical hypotheses**.  
These hypotheses are statements about a population parameter, in this case, the population mean body weight, denoted as $\mu$.

---

## **2. Key Concepts**

### **Null Hypothesis ($H_0$)**
This is a claim of **“no difference”** or **“no effect”** in the population. It represents the current or historical belief that we test against.

**Mathematically:**
$$
H_0: \mu = \mu_0
$$

where $\mu_0$ is the hypothesized population mean.

---

### **Alternative Hypothesis ($H_a$)**
This is a claim that $H_0$ is false — that there is some difference or effect.

Depending on the research question, it may be:

- **Two-sided (two-tailed):**  
  $$
  H_a: \mu \neq \mu_0
  $$  
  *(We suspect the mean has changed in either direction.)*

- **Right-sided:**  
  $$
  H_a: \mu > \mu_0
  $$  
  *(We suspect the mean is higher.)*

- **Left-sided:**  
  $$
  H_a: \mu < \mu_0
  $$  
  *(We suspect the mean is lower.)*

---

### **Goal of the Test**

Collect data and use sample evidence to challenge $H_0$ and possibly support $H_a$.  
In logic, this is a process of **deduction** ; reasoning from the assumption that $H_0$ is true and assessing whether the observed data are consistent with that assumption.

---

#  3. Illustrative Example: “Body Weight”

---

## **Problem Context**

In the 1970s, the mean body weight of U.S. men aged 20–29 was known to be:

$$\mu_0 = 170 \text{ pounds}$$

with a population standard deviation of:

$$\sigma = 40 \text{ pounds}$$

A current researcher suspects that the **average weight of today’s 20–29-year-old men has changed** (either increased or decreased).

A simple random sample (SRS) of:

$$n = 64$$

men in that age group yields a sample mean of:

$$\bar{x} = 185 \text{ pounds}$$

We will test whether this difference (185 vs. 170) is statistically significant.

---

## **Step A: Setting up Hypotheses**

$$
\begin{aligned}
H_0 &: \mu = 170 \quad &\text{(no change in mean weight)} \\
H_a &: \mu \ne 170 \quad &\text{(mean weight has changed)}
\end{aligned}
$$

**Interpretation:**

Under the null hypothesis, the true mean is assumed to be **170 lb**.  
If our observed mean (185 lb) lies far enough from 170, it may suggest that $H_0$ is unlikely to be true.

---

## **Step B: Identify Test Type**

We are performing a **one-sample z-test for a population mean**.  
This test is appropriate because the **population standard deviation ($\sigma$)** is known.

---

## **Step C: Test Statistic Formula**

$$
z_{\text{stat}} = \frac{\bar{x} - \mu_0}{SE_{\bar{x}}}
$$

where

$$
SE_{\bar{x}} = \frac{\sigma}{\sqrt{n}}
$$

is the **standard error of the mean**, representing the standard deviation of all possible sample means of size $n$.

---

## **Step D: Calculation**

Given:

$$\sigma = 40, \quad n = 64, \quad \bar{x} = 185, \quad \mu_0 = 170$$

Compute the standard error:

$$
SE_{\bar{x}} = \frac{40}{\sqrt{64}} = \frac{40}{8} = 5
$$

Now, compute the z-statistic:

$$
z_{\text{stat}} = \frac{185 - 170}{5} = \frac{15}{5} = 3.00
$$

---

## **Step E: Statistical Interpretation of z**

The value:

$$z = 3.00$$

indicates that the **observed sample mean (185 lb)** is **3 standard errors above** the hypothesized mean (170 lb).

**This means:**  
If the true mean were really 170 lb, observing a sample mean of 185 lb would be **very unlikely** — occurring only in the extreme right tail of the sampling distribution.

---

## **Step F: Distribution under $H_0$**

Under $H_0$:

$$
\bar{X} \sim N(\mu_0 = 170, \, SE_{\bar{X}} = 5)
$$

This implies:

- The expected sample means cluster around **170**.
- About 95% of all sample means would fall between:

$$
170 \pm 1.96(5) = 170 \pm 9.8 = (160.2, 179.8)
$$

The observed mean (**185 lb**) lies **beyond 179.8**, in the **outer 5% tail** of this distribution.

---

## **Step G: Immediate Statistical Implication**

Since:

$$z = 3.00$$

corresponds to a **p-value ≈ 0.0027 (two-sided)**, and

$$p < 0.05$$

we **reject the null hypothesis**.

---

## **Interpretation**

There is **strong evidence** that the **average weight of today’s 20–29-year-old men has changed** from the 1970s mean of 170 lb.

**In practical terms:**  
It has **likely increased**, reflecting a statistically significant upward shift in mean body weight.

---

## **4. P-Value and Statistical Reasoning**

The **p-value** answers:
> “What is the probability of obtaining a test statistic as extreme as, or more extreme than, the observed one, if $H_0$ is true?”

For a two-sided test:
$$
P = 2 \times P(Z > |z_{stat}|)
$$

**Calculation:**
From z-tables:
$$
P(Z > 3.00) = 0.0013
$$

Thus:
$$
P = 2 \times 0.0013 = 0.0026
$$

**Interpretation:**  
If the true mean body weight were still 170 lb, there would be only a **0.26%** chance of observing a sample mean as extreme as 185 lb by random variation alone.  
This is **strong evidence** that the true mean is not 170 lb.

---

# 5.  P-Value Formulas for Different Alternative Hypotheses

| **Alternative Hypothesis** | **P-Value Formula** | **Tail Area** |
|:----------------------------|:--------------------|:---------------|
| **$H_a : \mu > \mu_0$** | $$P = P(Z > z_{\text{stat}})$$ | **Right-tail** |
| **$H_a : \mu < \mu_0$** | $$P = P(Z < z_{\text{stat}})$$ | **Left-tail** |
| **$H_a : \mu \ne \mu_0$** | $$P = 2 \times (\text{one-tail area})$$ | **Two-tailed** |

----

## **6. One-Sided and Two-Sided P-Value Examples**

| $z_{stat}$ | One-sided $P$ | Two-sided $P$ | Interpretation |
|-------------|----------------|----------------|----------------|
| 0.60 | 0.2743 | 0.5486 | Non-significant — result consistent with $H_0$ |
| 3.00 | 0.0013 | 0.0026 | Highly significant — strong evidence against $H_0$ |

---

## **7. Interpretation of P-values**

A **small p-value** indicates that the observed data are **unlikely** under $H_0$, hence provide **strong evidence against** it.  
A **large p-value** suggests that the data are **compatible with $H_0$**.



| **P-value Range** | **Statistical Interpretation** | **Strength of Evidence Against $H_0$** |
|:------------------:|:------------------------------|:----------------------------------------|
| **$P > 0.10$** | The result is **non-significant**; fail to reject the null hypothesis. | **No evidence** against $H_0$ |
| **$0.05 < P \le 0.10$** | The result is **marginally significant**; may warrant further investigation. | **Weak evidence** against $H_0$ |
| **$0.01 < P \le 0.05$** | The result is **statistically significant**; reject the null hypothesis at the 5% level. | **Moderate evidence** against $H_0$ |
| **$P \le 0.01$** | The result is **highly significant**; reject the null hypothesis with strong confidence. | **Strong evidence** against $H_0$ |



For our test:  
$$
P = 0.0026 \le 0.01
$$  
 **Highly significant evidence against $H_0$.**

---

## **8. $\alpha$-Level (Significance Level)**

Let $\alpha$ be the **probability of incorrectly rejecting** $H_0$ when it is true (**Type I error**).

Common choices: $\alpha = 0.10, 0.05, 0.01$

**Decision Rule:**
- Reject $H_0$ if $P \le \alpha$
- Fail to reject $H_0$ if $P > \alpha$

**Example Decisions:**
- $\alpha = 0.10$, $P = 0.27$ ⇒ Fail to reject $H_0$
- $\alpha = 0.01$, $P = 0.001$ ⇒ Reject $H_0$
- Our case: $\alpha = 0.05$, $P = 0.0026$ ⇒ Reject $H_0$

**Interpretation:**  
There is **strong statistical evidence** that the mean weight of U.S. men aged 20–29 has **changed** since the 1970s.

## **9. (Summary) One-Sample z Test**

| **Step** | **Description** | **Formula** | **Example Value** | **Interpretation** |
|-----------|------------------|--------------|-------------------|--------------------|
| A | Hypothesis statements | $H_0: \mu = 170$, $H_a: \mu \neq 170$ | — | Two-tailed test for change |
| B | Test statistic | $z = \dfrac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$ | $z = 3.00$ | Sample mean 3 SEs above $\mu_0$ |
| C | P-value | $2 \times P(Z > 3.00) = 0.0026$ | — | Very unlikely under $H_0$ |
| D | Decision | Reject $H_0$ if $P \le 0.05$ | $P = 0.0026 < 0.05$ | Reject $H_0$ |
| E | Conclusion | Mean body weight has changed | — | Evidence of weight increase |

---

## **10. Conditions for Valid z-test**

The results are statistically valid if:

1. $\sigma$ (population standard deviation) is known.  
2. The sample is random and independent.  
3. The population is approximately Normal **or** $n \ge 30$ (Central Limit Theorem).  
4. Data are reliable and accurately measured.

 All these conditions are satisfied in our example, hence the inference is **valid**.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -------------------------------------------
# GIVEN VALUES
# -------------------------------------------
mu0 = 170       # Null mean
sigma = 40      # Known population SD
n = 64          # Sample size
xbar = 185      # Sample mean
alpha = 0.05    # Significance level

# Hypothesized shift for H1
mu1 = xbar      # Alternative mean = observed mean

# -------------------------------------------
# COMPUTE Z-STATISTIC AND P-VALUE
# -------------------------------------------
se = sigma / np.sqrt(n)
z_stat = (xbar - mu0) / se
p_value = 2 * norm.sf(abs(z_stat))
z_crit = norm.ppf(1 - alpha/2)

decision = "Reject H₀" if p_value <= alpha else "Fail to Reject H₀"

# -------------------------------------------
# POWER CALCULATION
# -------------------------------------------
delta = (mu1 - mu0) / se
power = norm.sf(z_crit - delta) + norm.cdf(-z_crit - delta)
beta = 1 - power

# -------------------------------------------
# PRINT NUMERICAL OUTPUT
# -------------------------------------------
print("===== One-Sample Z-Test (Body Weight) =====")
print(f"Hypothesized Mean (μ₀): {mu0}")
print(f"Sample Mean (x̄): {xbar}")
print(f"Population SD (σ): {sigma}")
print(f"Sample Size (n): {n}")
print(f"Standard Error (SE): {se:.2f}")
print(f"z-Statistic: {z_stat:.2f}")
print(f"p-Value (two-sided): {p_value:.4f}")
print(f"Decision at α={alpha}: {decision}")
print("\n===== Power Analysis =====")
print(f"Assumed True Mean (μ₁): {mu1}")
print(f"Noncentrality (δ): {delta:.3f}")
print(f"Power (1-β): {power:.3f}")
print(f"Type II Error (β): {beta:.3f}")
print("====================================================")
```

The sample mean is substantially higher than the hypothesized mean, producing a large $z$-statistic and a very small $p$-value. Therefore, we reject the null hypothesis at the $5\%$ significance level. The test also has high power ($\approx 85\%$), indicating strong ability to detect this difference, with low Type II error risk.


```python
# -------------------------------------------
# VISUALIZATIONS
# -------------------------------------------
x = np.linspace(-4, 4, 400)

fig, axs = plt.subplots(1, 3, figsize=(22, 6))

# ----- Plot 1: Sampling Distribution Under H0 -----
axs[0].plot(x, norm.pdf(x), color="black", linewidth=2, label="H₀ Distribution")
axs[0].fill_between(x, norm.pdf(x), where=(x > -z_crit) & (x < z_crit),
                    color="lightblue", alpha=0.7, label="Fail to Reject H₀")
axs[0].fill_between(x, norm.pdf(x), where=(x <= -z_crit) | (x >= z_crit),
                    color="lightcoral", alpha=0.7, label="Reject H₀")
axs[0].axvline(z_stat, color="blue", linestyle="--", linewidth=2, label=f"Observed z = {z_stat:.2f}")
axs[0].set_title("Sampling Distribution Under H₀")
axs[0].set_xlabel("z-value")
axs[0].set_ylabel("Density")
axs[0].legend()

# ----- Plot 2: Sampling Distribution Under H1 -----
axs[1].plot(x, norm.pdf(x - delta), color="green", linewidth=2, label="H₁ Distribution")
axs[1].axvline(z_crit, color="red", linestyle="--", linewidth=2, label="Critical Values")
axs[1].axvline(-z_crit, color="red", linestyle="--", linewidth=2)
axs[1].set_title("Sampling Distribution Under H₁ (Mean Shifted)")
axs[1].set_xlabel("z-value")
axs[1].set_ylabel("Density")
axs[1].legend()

# ----- Plot 3: Power Visualization -----
axs[2].plot(x, norm.pdf(x), color="blue", linewidth=2, label="H₀ Distribution")
axs[2].plot(x, norm.pdf(x - delta), color="green", linestyle="--", linewidth=2, label="H₁ Distribution")
axs[2].fill_between(x, norm.pdf(x - delta), where=(x <= -z_crit) | (x >= z_crit),
                    color="lightgreen", alpha=0.7, label="Power (Correct Reject)")
axs[2].fill_between(x, norm.pdf(x - delta), where=(x > -z_crit) & (x < z_crit),
                    color="yellow", alpha=0.6, label="Type II Error (β)")
axs[2].set_title("Power and Type II Error Visualization")
axs[2].set_xlabel("z-value")
axs[2].set_ylabel("Density")
axs[2].legend()

plt.tight_layout()
plt.show()
```

# **The Lake Wobegon Example**

> “Where all the children are above average”

---

## **Overview**

This is a one-sample $z$-test for a population mean, where the population standard deviation $\sigma$ is known.

---

## **Background and Context**

In this fictional town of **Lake Wobegon**, it’s said that “all the children are above average.”  
We’ll test this claim statistically using IQ scores measured by the **Wechsler Adult Intelligence Scale (WAIS)**.

Let $X$ represent WAIS IQ scores.

Typically, for the general population:

$$
X \sim N(\mu = 100, \sigma = 15)
$$

where:

- $\mu = 100$: population mean IQ  
- $\sigma = 15$: population standard deviation

We take a simple random sample (SRS) of $n = 9$ adults from Lake Wobegon, obtaining:

$$
\{116, 128, 125, 119, 89, 99, 105, 116, 118\}
$$

---

## **Sample Calculation**

$$
\bar{x} = \frac{116 + 128 + 125 + 119 + 89 + 99 + 105 + 116 + 118}{9} = 112.8
$$

**Interpretation:**  
The average IQ of the 9 sampled residents is **112.8**, which appears higher than the national average ($100$).  
We now test whether this observed difference is statistically significant.

---

## **A. Hypotheses**

We formalize the research question:  
Is the mean IQ in Lake Wobegon greater than 100?

$$
H_0: \mu = 100 \quad \text{(no difference — average IQ same as national mean)}
$$

$$
H_a: \mu > 100 \quad \text{(mean IQ in Lake Wobegon is higher)}
$$

**Interpretation:**  
The null hypothesis $H_0$ assumes the town’s mean IQ is not special — same as everyone else’s.  
The alternative $H_a$ represents the claim that “Lake Wobegon residents are above average,” implying a one-sided (right-tailed) test.

---

## **B. Test Statistic**

We use the $z$-statistic since $\sigma$ (the population standard deviation) is known.

$$
z_{stat} = \frac{\bar{x} - \mu_0}{SE_{\bar{x}}}
$$

where the standard error of the mean is

$$
SE_{\bar{x}} = \frac{\sigma}{\sqrt{n}}
$$

**Calculation:**

$$
SE_{\bar{x}} = \frac{15}{\sqrt{9}} = \frac{15}{3} = 5
$$

Then:

$$
z_{stat} = \frac{112.8 - 100}{5} = \frac{12.8}{5} = 2.56
$$

**Interpretation:**  
The sample mean of 112.8 is **2.56 standard errors** above the hypothesized population mean of 100.  
If the true mean were indeed 100, observing such a large deviation would be relatively rare — which hints that $H_0$ may not be true.

---

## **C. P-value Calculation**

The P-value quantifies how likely it is to observe such a result (or more extreme) under the null hypothesis.

Since this is a right-tailed test ($H_a: \mu > 100$):

$$
P = P(Z \ge z_{stat})
$$

From the standard normal distribution table:

$$
P(Z \ge 2.56) = 0.0052
$$

**Interpretation:**  
If the true mean IQ were 100, the probability of getting a sample mean at least as large as 112.8 (or equivalently, $z = 2.56$) is only **0.52%**.  
This is very unlikely under $H_0$, indicating strong statistical evidence that the true mean IQ is indeed greater than 100.

---

## **D. Decision (One-Sided Test)**

We compare the P-value to a chosen significance level (say $\alpha = 0.05$):

$$
P = 0.0052 < 0.05
$$

 **Decision:** Reject $H_0$

**Interpretation:**  
At the 5% level of significance, there is strong evidence that the mean IQ in Lake Wobegon exceeds 100.  
Statistically, this supports the town’s “above-average” reputation!

---

## **E. Two-Sided Version (for Comparison)**

If instead, we were testing:

$$
H_0: \mu = 100 \quad \text{vs.} \quad H_a: \mu \ne 100
$$

then the P-value doubles, because both tails of the distribution are considered:

$$
P_{two-sided} = 2 \times 0.0052 = 0.0104
$$

**Interpretation:**  
Even under a two-tailed test, $P = 0.0104 < 0.05$, so we would still reject $H_0$.  
Hence, whether we test for “greater than” or “different from,” the evidence points to a true mean above 100.

---

## **F. Statistical Reasoning and Interpretation**

| **Symbol** | **Meaning** | **Lake Wobegon Example** | **Interpretation** |
|-------------|--------------|---------------------------|---------------------|
| $\mu_0$ | Hypothesized population mean | 100 | Reference “average IQ” |
| $\sigma$ | Population SD | 15 | Natural variation in IQ scores |
| $n$ | Sample size | 9 | Number of residents sampled |
| $\bar{x}$ | Sample mean | 112.8 | Observed sample IQ average |
| $SE_{\bar{x}}$ | Standard error | 5 | Expected variability of sample means |
| $z_{stat}$ | Test statistic | 2.56 | 2.56 SEs above hypothesized mean |
| $P$ | P-value | 0.0052 | Chance of seeing such data if $H_0$ true |
| $\alpha$ | Significance level | 0.05 | Threshold for decision |
| — | Decision | Reject $H_0$ | Mean IQ significantly higher |

---

## **G. Interpretation**

**Evidence strength:**  
$P = 0.0052$ is much less than 0.05 → strong evidence that $\mu > 100$.

**Conclusion:**  
The Lake Wobegon sample provides statistically significant evidence that the town’s population mean IQ is higher than the general population’s.

**In plain words:**  
It’s highly unlikely (probability < 1%) that this sample came from a population with mean 100.  
Therefore, the claim that “all the children are above average” — while exaggerated — is **statistically supported by the data.**

---

## **H. Assumptions Check**

For the z-test results to be valid, the following assumptions must hold:

1. **Population standard deviation known:** $\sigma = 15$ (given by WAIS).  
2. **Simple Random Sample (SRS):** The 9 individuals were randomly chosen.  
3. **Approximate Normality:** Since $X \sim N(100, 15)$ by definition, the sampling distribution of $\bar{X}$ is normal.  
4. **Independence:** Each observation represents a different person.

**All conditions are satisfied**, so the inference is reliable.

---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -------------------------------------------
# DATA
# -------------------------------------------
data = np.array([116,128,125,119,89,99,105,116,118])
mu0 = 100          # Hypothesized mean
sigma = 15         # Known population SD
alpha = 0.05       # Significance level
n = len(data)      # Sample size
xbar = np.mean(data)

# -------------------------------------------
# Z-TEST (ONE-SIDED: Ha: μ > μ0)
# -------------------------------------------
se = sigma / np.sqrt(n)
z_stat = (xbar - mu0) / se
p_value = norm.sf(z_stat)  # right-tailed

decision = "Reject H₀" if p_value <= alpha else "Fail to Reject H₀"

# -------------------------------------------
# POWER CALCULATION (Assume true mean = x̄)
# -------------------------------------------
mu1 = xbar
delta = (mu1 - mu0) / se   # noncentrality shift
z_crit = norm.ppf(1 - alpha)
power = norm.sf(z_crit - delta)
beta = 1 - power

# -------------------------------------------
# PRINT RESULTS
# -------------------------------------------
print("===== One-Sample Right-Tailed Z-Test (Lake Wobegon IQ) =====")
print(f"Sample Size (n): {n}")
print(f"Population SD (σ): {sigma}")
print(f"Sample Mean (x̄): {xbar:.2f}")
print(f"Hypothesized Mean (μ₀): {mu0}")
print(f"Standard Error (SE): {se:.2f}")
print(f"z-Statistic: {z_stat:.2f}")
print(f"P-Value (right-tailed): {p_value:.4f}")
print(f"Decision at α={alpha}: {decision}")
print("\n===== Power Analysis =====")
print(f"Assumed True Mean (μ₁): {mu1:.2f}")
print(f"Noncentrality (δ): {delta:.3f}")
print(f"Power (1-β): {power:.3f}")
print(f"Type II Error (β): {beta:.3f}")
print("==============================================================")
```

The sample mean is significantly higher than the hypothesized population mean, yielding a large positive z-value and a small right-tailed p-value. Therefore, we reject the null hypothesis and conclude that mean IQ in Lake Wobegon is likely above average. The test has strong power, indicating a low probability of Type II error.

```python
# -------------------------------------------
# VISUALIZATIONS
# -------------------------------------------
x = np.linspace(-4, 4, 400)

fig, axs = plt.subplots(1, 3, figsize=(22, 6))

# ----- Plot 1: H0 Distribution -----
axs[0].plot(x, norm.pdf(x), color="black", linewidth=2, label="H₀ Distribution")
axs[0].fill_between(x, norm.pdf(x), where=(x <= z_crit),
                    color="lightblue", alpha=0.7, label="Fail to Reject H₀")
axs[0].fill_between(x, norm.pdf(x), where=(x > z_crit),
                    color="lightcoral", alpha=0.7, label="Reject H₀")
axs[0].axvline(z_stat, color="blue", linestyle="--", linewidth=2, label=f"Observed z = {z_stat:.2f}")
axs[0].set_title("Sampling Distribution Under H₀")
axs[0].set_xlabel("z-value")
axs[0].set_ylabel("Density")
axs[0].legend()

# ----- Plot 2: H1 Distribution (Shifted) -----
axs[1].plot(x, norm.pdf(x - delta), color="green", linewidth=2, label="H₁ Distribution")
axs[1].axvline(z_crit, linestyle="--", color="red", linewidth=2, label="Critical Value")
axs[1].set_title("Sampling Distribution Under H₁")
axs[1].set_xlabel("z-value")
axs[1].set_ylabel("Density")
axs[1].legend()

# ----- Plot 3: Power vs Type II Error -----
axs[2].plot(x, norm.pdf(x), color="blue", linewidth=2, label="H₀ Distribution")
axs[2].plot(x, norm.pdf(x - delta), color="green", linestyle="--", linewidth=2, label="H₁ Distribution")
axs[2].fill_between(x, norm.pdf(x - delta), where=(x > z_crit),
                    color="lightgreen", alpha=0.7, label="Power (Reject H₀)")
axs[2].fill_between(x, norm.pdf(x - delta), where=(x <= z_crit),
                    color="yellow", alpha=0.6, label="Type II Error (β)")
axs[2].set_title("Power and β Visualization")
axs[2].set_xlabel("z-value")
axs[2].set_ylabel("Density")
axs[2].legend()

plt.tight_layout()
plt.show()
```

#  Power and Sample Size

Statistical inference doesn’t just end when you *“reject or retain”* the null hypothesis.  
We must also understand how **reliable** the test is — i.e., how likely it is to correctly detect a true effect when it exists.  
This introduces the concepts of **Type I and Type II errors**, and the **power** of a test.

---

## A. Decision Errors in Hypothesis Testing

Whenever we make a decision based on a sample, two types of errors can occur:

| Truth (Population) | Decision (Sample) | Outcome | Probability Symbol |
|--------------------|------------------|----------|--------------------|
| $H_0$ true | Retain $H_0$ |  Correct retention | — |
| $H_0$ true | Reject $H_0$ |  Type I error | $\alpha$ |
| $H_0$ false | Retain $H_0$ | Type II error | $\beta$ |
| $H_0$ false | Reject $H_0$ | Correct rejection | $1 - \beta$ |

---

## B. Interpreting the Two Types of Error

**Type I error ($\alpha$):**  
Rejecting a true null hypothesis (a “false positive”).  
Example: Concluding that Lake Wobegon IQ is higher when it’s actually equal to 100.

**Type II error ($\beta$):**  
Failing to reject a false null hypothesis (a “false negative”).  
Example: Concluding that Lake Wobegon IQ is average when it’s actually higher.

---

## C. Statistical Power

The power of a test measures its sensitivity — the probability of detecting a true effect when one exists.

$$
1 - \beta = P(\text{Reject } H_0 \mid H_0 \text{ is false})
$$

**In words:**  
Power is the probability that the test correctly rejects $H_0$ when the alternative hypothesis $H_a$ is true.

- High power (close to 1) → the test is good at finding true differences.  
- Low power → the test often “misses” real effects.

---

## D. Formula for Power of a One-Sample z-Test

When population standard deviation $\sigma$ is known, and we test:

$$
H_0 : \mu = \mu_0 \quad \text{vs.} \quad H_a : \mu = \mu_a
$$

the power of the z-test is:

$$
1 - \beta = \Phi\left(-z_{1-\frac{\alpha}{2}} + \frac{|\mu_a - \mu_0|\sqrt{n}}{\sigma}\right)
$$

where:

- $\Phi(z)$: cumulative probability from the standard normal distribution  
- $z_{1-\frac{\alpha}{2}}$: critical value for the chosen significance level (two-tailed test)  
- $\mu_0$: mean under $H_0$  
- $\mu_a$: mean under $H_a$  
- $n$: sample size  
- $\sigma$: population standard deviation

---

## E. Example: Calculating Power

**Scenario:**  
A study with $n = 16$ tests

$$
H_0: \mu = 170
$$

at $\alpha = 0.05$ (two-sided), with $\sigma = 40$.  
We want to find the power to detect a true population mean of $\mu_a = 190$.

### Step 1: Identify known quantities
| Parameter | Value |
|------------|--------|
| $\mu_0$ | 170 |
| $\mu_a$ | 190 |
| $\sigma$ | 40 |
| $n$ | 16 |
| $\alpha$ | 0.05 (two-sided) |
| $z_{1-\alpha/2}$ | 1.96 |

---

### Step 2: Substitute into the power formula

$$
1 - \beta = \Phi\left(-z_{1-\frac{\alpha}{2}} + \frac{|\mu_a - \mu_0|\sqrt{n}}{\sigma}\right)
$$

Compute each part:

$$
|\mu_a - \mu_0| = |190 - 170| = 20
$$

$$
\frac{|\mu_a - \mu_0|\sqrt{n}}{\sigma} = \frac{20 \times 4}{40} = 2
$$

Now substitute:

$$
1 - \beta = \Phi(-1.96 + 2) = \Phi(0.04)
$$

From normal tables:

$$
\Phi(0.04) = 0.5160
$$

---

### Step 3: Interpretation

$$
1 - \beta = 0.5160
$$

The test has a **power of 0.516**, or **51.6%**, to detect a true mean of 190.

**Interpretation:**  
If the true mean were 190, there would be only a 51.6% chance of correctly rejecting $H_0$.  
To improve detection, we would need to increase the sample size.

---

## F. Understanding Power Graphically

Imagine two sampling distributions of $\bar{X}$:

- Top curve: under $H_0 : \mu = 170$  
- Bottom curve: under $H_a : \mu = 190$

We set $\alpha = 0.05$ (two-sided).  
$H_0$ is rejected for large sample means (in the tails).

The right-tail cutoff corresponds to $\bar{x} = 189.6$.  
Under $H_a$ (true mean = 190), the probability that $\bar{X} > 189.6$ is 0.5160 — the **power** of the test.

---

## G. Increasing Power

Power increases when:

- $n$ (sample size) ↑  
- $\sigma$ (population variability) ↓  
- $|\mu_a - \mu_0|$ (true difference) ↑  
- $\alpha$ (significance level) ↑ — though this also increases Type I error

---

## H. Determining Sample Size for Desired Power

To plan a study, we often want to know how large a sample is needed to achieve a specified power.

$$
n = \left( \frac{\sigma (z_{1-\frac{\alpha}{2}} + z_{1-\beta})}{\Delta} \right)^2
$$

where:

- $\Delta = |\mu_a - \mu_0|$: the difference worth detecting  
- $1 - \beta$: desired power  
- $\alpha$: significance level  
- $\sigma$: population standard deviation

---

## I. Example: Required Sample Size

**Goal:** Find the sample size required for a one-sample z-test with:

| Parameter | Value |
|------------|--------|
| Desired power | $1 - \beta = 0.90$ |
| $\alpha$ | 0.05 (two-tailed) |
| $\sigma$ | 40 |
| $\mu_0$ | 170 |
| $\mu_a$ | 190 |
| $\Delta = |\mu_a - \mu_0|$ | 20 |

### Step 1: Identify critical z-values
$$
z_{1-\frac{\alpha}{2}} = 1.96 \quad (\alpha = 0.05)
$$
$$
z_{1-\beta} = 1.28 \quad (\text{for power } = 0.90)
$$

### Step 2: Substitute into the formula
$$
n = \left( \frac{40(1.96 + 1.28)}{20} \right)^2
$$

Simplify:
$$
n = (6.48)^2 = 41.99
$$

Thus:
$$
n = 42
$$

---

### Step 3: Interpretation

To achieve **90% power** at a **5% significance level (two-sided)**, we need a sample of at least **42 observations**.

With $n=16$, we only had **51.6% power**; with $n=42$, power increases to **90%**.

---

## J. Summary of Key Relationships

| Factor | Effect on Power | Explanation |
|---------|----------------|--------------|
| ↑ Sample size ($n$) | ↑ Power | Reduces sampling variability |
| ↓ Standard deviation ($\sigma$) | ↑ Power | Less noise improves detection |
| ↑ True difference ($|\mu_a - \mu_0|$) | ↑ Power | Larger effect easier to detect |
| ↑ Significance level ($\alpha$) | ↑ Power (but ↑ Type I error) | Wider rejection region |

---

##  Final Takeaway

Power quantifies how likely we are to detect a real effect.  
Sample size planning ensures that our study has enough data to detect meaningful differences.  
Low power → high risk of **Type II errors** (missing true differences).  

**Statistical rigor requires balancing $\alpha$, $\beta$, and $n$ to design reliable studies.**


```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# ===============================
# POWER AND SAMPLE SIZE FUNCTIONS
# ===============================

def power_one_sample_z(mu0, mua, sigma, n, alpha=0.05, alternative='two-sided'):
    """
    Computes the statistical power of a one-sample z-test.

    Parameters:
        mu0 (float): Hypothesized population mean under H0
        mua (float): True population mean under Ha
        sigma (float): Population standard deviation
        n (int): Sample size
        alpha (float): Significance level
        alternative (str): 'two-sided', 'greater', or 'less'

    Returns:
        power (float): Probability of rejecting H0 when Ha is true
    """
    SE = sigma / np.sqrt(n)
    delta = (mua - mu0) / SE  # noncentrality

    if alternative == 'two-sided':
        z_crit = norm.ppf(1 - alpha/2)
        power = norm.cdf(-z_crit + delta) + (1 - norm.cdf(z_crit + delta))

    elif alternative == 'greater':
        z_crit = norm.ppf(1 - alpha)
        power = 1 - norm.cdf(z_crit - delta)

    elif alternative == 'less':
        z_crit = norm.ppf(alpha)
        power = norm.cdf(z_crit - delta)

    return power


def required_sample_size(mu0, mua, sigma, alpha=0.05, power=0.80):
    """
    Computes the sample size required for a one-sample z-test to achieve a given power.
    Uses the standard normal approximation.
    """
    delta = abs(mua - mu0)
    z_alpha = norm.ppf(1 - alpha/2)
    z_beta = norm.ppf(power)
    n = (sigma * (z_alpha + z_beta) / delta) ** 2
    return int(np.ceil(n))


# ===============================
# EXAMPLE PARAMETERS
# ===============================
mu0 = 170          # Null hypothesis mean
mua = 190          # True mean under Ha
sigma = 40         # Population SD
alpha = 0.05       # Significance level

# Compute example power for n = 16
n_example = 16
example_power = power_one_sample_z(mu0, mua, sigma, n_example, alpha, alternative='two-sided')

# Compute required sample size for 90% power
target_power = 0.90
n_required = required_sample_size(mu0, mua, sigma, alpha, target_power)

print(f"Power at n={n_example}: {example_power:.3f}")
print(f"Required sample size for {target_power*100:.0f}% power: n = {n_required}")


# ===============================
# POWER CURVE PLOT
# ===============================
sample_sizes = np.arange(5, 101)
power_values = [power_one_sample_z(mu0, mua, sigma, n, alpha) for n in sample_sizes]

plt.figure(figsize=(10, 5))
plt.plot(sample_sizes, power_values, linewidth=2)
plt.axhline(target_power, linestyle='--', label=f"Desired Power = {target_power}")
plt.axvline(n_required, linestyle='--', label=f"Required n = {n_required}")
plt.xlabel("Sample Size (n)")
plt.ylabel("Power (1 - β)")
plt.title("Power Curve for One-Sample Z-Test")
plt.legend()
plt.grid(True)
plt.show()
```
