# 4.1 Discrete Uniform Distribution

---

###  Definition

A **discrete random variable** $X$ is said to follow a **discrete uniform distribution** if it takes on a finite set of values:

$$
\{x_1, x_2, \dots, x_k\}
$$

with **equal probability** for each outcome.

---

###  Probability Mass Function (PMF)

The PMF defines the probability that $X$ takes a particular value $x$.  
For the discrete uniform distribution:

$$
f(x) = P(X = x) =
\begin{cases}
\dfrac{1}{k}, & x \in \{x_1, x_2, \dots, x_k\} \\
0, & \text{otherwise}
\end{cases}
$$

---

> “The probability that $X$ equals $x$ is $\frac{1}{k}$ for all allowed values of $x$; otherwise, it is 0.”
---

### Notation

- $X$ — the random variable, representing the outcome of a random experiment.  
- $x$ — a specific value that $X$ can take.  
- $P(X = x)$ — the probability that $X$ equals a specific value $x$.  
- $k$ — the total number of distinct values that $X$ can assume.  
- $f(x)$ — shorthand for the **probability mass function (PMF)**, which maps each possible outcome $x$ to its probability.

---

###  Key Properties

#### 1. **Equally Likely Outcomes**

Each possible outcome occurs with probability:

$$
P(X = x) = \frac{1}{k}
$$

---

#### 2. **Expected Value (Mean)**

$$
E[X] = \frac{1}{k} \sum_{i=1}^{k} x_i
$$

This represents the **long-run average** outcome if the experiment is repeated infinitely many times.

---

#### 3. **Variance**

$$
Var(X) = \frac{1}{k} \sum_{i=1}^{k} (x_i - E[X])^2
$$

This measures the **average squared deviation** of outcomes from the mean.

---

##  Worked Example: Rolling a Fair Die

---

### Step 1: Define the Random Variable

Let $X$ denote the number observed when rolling a fair six-sided die:

$$
X \in \{1, 2, 3, 4, 5, 6\}
$$

---

### Step 2: Probability Mass Function

$$
f(x) = P(X = x) =
\begin{cases}
\dfrac{1}{6}, & x = 1, 2, 3, 4, 5, 6 \\
0, & \text{otherwise}
\end{cases}
$$

**Interpretation:**  
Each number from 1 to 6 has an equal probability of $\frac{1}{6}$.  
Any number not in this range has a probability of 0.

---

### Step 3: Expected Value

$$
E[X] = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = \frac{21}{6} = 3.5
$$

**Interpretation:**  
If we roll the die many times, the average observed value tends toward **3.5**.

---

### Step 4: Variance

$$
Var(X) = \frac{1}{6}\sum_{x=1}^{6}(x - 3.5)^2
$$

Compute step-by-step:

$$
Var(X) = \frac{1}{6}(6.25 + 2.25 + 0.25 + 0.25 + 2.25 + 6.25)
$$

$$
Var(X) = \frac{17.5}{6} \approx 2.9167
$$

Then, the **standard deviation** is:

$$
\sigma_X = \sqrt{Var(X)} \approx 1.71
$$

**Interpretation:**  
On average, die outcomes deviate from the mean (3.5) by approximately **1.71**.

---

### Step 5: Practical Significance

- The **expected value** lies exactly at the **midpoint** of the range, a hallmark of **symmetric uniform distributions**.  
- The **variance** quantifies the spread of outcomes, showing how dispersed results are from the mean.  
- Discrete uniform distributions are foundational in:
  - **Simulation studies**
  - **Monte Carlo experiments**
  - **Random sampling**
  - **Algorithm testing**

where all outcomes are **equally likely by design**.

---
### Summary Table — Discrete Uniform Distribution  

| **Concept** | **Formula** | **Description** |
|:-------------|:-------------|:----------------|
| **Support** | $\{x_1, x_2, \dots, x_k\}$ | Finite set of equally likely outcomes. |
| **Probability Mass Function (PMF)** | $f(x) = 1/k$ | Each outcome has equal probability $1/k$. |
| **Expected Value** | $E[X] = \frac{1}{k} \sum_{i=1}^{k} x_i$ | Theoretical long-run mean of the outcomes. |
| **Variance** | $Var(X) = \frac{1}{k} \sum_{i=1}^{k} (x_i - E[X])^2$ | Average squared deviation from the mean. |
| **Standard Deviation** | $\sigma_X = \sqrt{Var(X)}$ | Typical amount by which $X$ deviates from its mean. |
| **Moment Generating Function (MGF)** | $M_X(t) = \frac{1}{k} \sum_{i=1}^{k} e^{t x_i}$ | Encodes all the moments of the distribution. |
| **Entropy** | $H(X) = \log(k)$ | Measure of uncertainty; maximum when all outcomes are equally likely. |
| **Example (Fair Die)** | $E[X] = 3.5$, $Var(X) \approx 2.92$, $\sigma_X \approx 1.71$ | Six equally likely outcomes: $\{1, 2, 3, 4, 5, 6\}$. |
| **Shape** | — | Uniform (flat) — all outcomes equally probable. |
| **Applications** | — | Modeling fair games, lotteries, or random selection from a finite set. |



---

###  Key Takeaway

The **discrete uniform distribution** forms the backbone of **randomness** in statistics and data science.  
It models perfectly **fair** and **unbiased** scenarios, providing a theoretical foundation for **simulation**, **sampling**, and **probabilistic modeling**.

---



```python
# ==========================================================
# 4.1 Discrete Uniform Distribution
# ==========================================================

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ----------------------------------------------------------
# Step 1: Define the Random Variable and Parameters
# ----------------------------------------------------------
# Random variable X: represents the outcomes of a fair six-sided die
X = np.arange(1, 7)            # Possible outcomes {1, 2, 3, 4, 5, 6}
k = len(X)                     # Total number of outcomes
pmf = np.ones(k) / k           # Probability Mass Function (PMF) – equal probabilities

# ----------------------------------------------------------
# Step 2: Compute Theoretical Statistics
# ----------------------------------------------------------
E_X = np.sum(X * pmf)                          # Theoretical Mean (Expected Value)
Var_X = np.sum((X - E_X)**2 * pmf)             # Theoretical Variance
Std_X = np.sqrt(Var_X)                         # Standard Deviation

# ----------------------------------------------------------
# Step 3: Monte Carlo Simulation
# ----------------------------------------------------------
n_sim = 100_000                                # Number of random simulations
samples = np.random.choice(X, size=n_sim, replace=True)
emp_mean = np.mean(samples)                    # Empirical mean from simulations
emp_var = np.var(samples)                      # Empirical variance

# ----------------------------------------------------------
# Step 4: Display Analytical Results
# ----------------------------------------------------------
print("=== DISCRETE UNIFORM DISTRIBUTION: FAIR DIE ===")
print("Support (X):", X)
print("PMF: f(x) = 1/k for x in {1,...,6}, else 0")
print(f"k = {k}")
print(f"Theoretical Mean (E[X])       = {E_X:.4f}")
print(f"Theoretical Variance (Var[X]) = {Var_X:.4f}")
print(f"Standard Deviation (σ[X])     = {Std_X:.4f}")
print(f"Empirical Mean (Simulation)   = {emp_mean:.4f}")
print(f"Empirical Variance            = {emp_var:.4f}")
print("------------------------------------------------------------\n")

# Generate PMF and CDF Table
cdf = np.cumsum(pmf)
df = pd.DataFrame({
    "x": X,
    "P(X = x)": pmf,
    "F(x) = P(X ≤ x)": cdf
})
display(df)

# ----------------------------------------------------------
# Step 5: Visualization
# ----------------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
fig.suptitle("Discrete Uniform Distribution (Fair Die): PMF | CDF | Simulation",
             fontsize=14, fontweight='bold')

# --- Plot 1: PMF ---
axes[0].bar(X, pmf, color='royalblue', edgecolor='black')
axes[0].set_title("Probability Mass Function (PMF)", fontsize=12)
axes[0].set_xlabel("x (Outcome)")
axes[0].set_ylabel("P(X = x)")
axes[0].set_xticks(X)
axes[0].grid(axis='y', linestyle=':', linewidth=0.6)

# --- Plot 2: CDF ---
axes[1].step(X, cdf, where='mid', color='crimson', linewidth=2)
axes[1].set_title("Cumulative Distribution Function (CDF)", fontsize=12)
axes[1].set_xlabel("x (Outcome)")
axes[1].set_ylabel("F(x) = P(X ≤ x)")
axes[1].set_xticks(X)
axes[1].grid(axis='both', linestyle=':', linewidth=0.6)

# --- Plot 3: Monte Carlo Simulation Histogram ---
axes[2].hist(samples, bins=np.arange(0.5, 7.5, 1),
             density=True, alpha=0.6, color='skyblue',
             edgecolor='black', label='Empirical')
axes[2].plot(X, pmf, 'ro--', label='Theoretical', linewidth=2)
axes[2].set_title(f"Monte Carlo Simulation ({n_sim:,} Trials)", fontsize=12)
axes[2].set_xlabel("x (Outcome)")
axes[2].set_ylabel("Probability")
axes[2].set_xticks(X)
axes[2].legend()
axes[2].grid(axis='y', linestyle=':', linewidth=0.6)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# ----------------------------------------------------------
# Step 6: Interpretation
# ----------------------------------------------------------
print("Interpretation:")
print("- The PMF is uniform, meaning all outcomes have equal probability (1/6).")
print("- The CDF increases linearly, showing equal incremental probability accumulation.")
print("- Monte Carlo simulation confirms the theoretical results — empirical frequencies converge to 1/6.")
print("- The mean (3.5) represents the center of the distribution, while the variance (~2.92) indicates the spread.")
print("- This model applies to fair dice, random integer generators, and uniform discrete sampling.")
```

# 4.2 Binomial Distribution  

---

### **Definition**

The **binomial distribution** models the **number of successes** in a fixed number of **independent Bernoulli trials**, each with the same probability of success $p$.

---

## 1️⃣ Bernoulli Trial and Bernoulli Process

### **Bernoulli Trial**

A **Bernoulli trial** is an experiment with exactly **two possible outcomes**:

- **Success (s):** occurs with probability $p$  
- **Failure (f):** occurs with probability $q = 1 - p$

**Examples:**

| Experiment | Success ($s$) | Failure ($f$) |
|-------------|----------------|----------------|
| Coin toss | Heads | Tails |
| Quality inspection | Defective | Non-defective |
| Customer response | Buys | Does not buy |

---

### **Bernoulli Process**

A **Bernoulli process** is a sequence of $n$ independent Bernoulli trials where:

- Each trial has the **same probability of success** $p$  
- Trials are **independent** — one trial’s outcome does not affect another  

---

## 2️⃣ Binomial Random Variable

Let $X$ denote the **number of successes** in $n$ Bernoulli trials.

**Notation:**
$$
X \sim \text{Binomial}(n, p)
$$

**Interpretation:**  
> “$X$ counts the number of successes in $n$ trials, each with success probability $p$.”

---

## 3️⃣ Probability Mass Function (PMF)

The probability of observing **exactly $x$ successes** is given by:

$$
f(x) = P(X = x) =
\begin{cases}
\displaystyle \binom{n}{x} p^{x} (1 - p)^{n - x}, & x = 0, 1, 2, \dots, n \\
0, & \text{otherwise}
\end{cases}
$$

---

###  **Notations**

- $\displaystyle \binom{n}{x}$ — *“The number of ways to choose $x$ successes from $n$ trials.”*  
  Think of it as counting arrangements — e.g., 3 heads in 5 coin tosses can occur as **HHTTT**, **HTHTT**, **TTHHH**, etc.

- $p^{x}$ — *“The probability of success raised to the number of successes.”*  
  Each success contributes a factor of $p$.

- $(1-p)^{n-x}$ — *“The probability of failure raised to the number of failures.”*

---

**Combined Reading:**

> “The probability of observing exactly $x$ successes in $n$ independent trials equals  
> the number of possible success arrangements times the probability of each specific arrangement.”

---

### **Zero Probability Elsewhere**

$$
f(x) = 0, \quad \text{for any } x \notin \{0, 1, 2, \dots, n\}
$$

This makes intuitive sense — you can’t have more successes than the number of trials.

---

## 4️⃣ Expected Value and Variance

For $X \sim \text{Binomial}(n, p)$:

$$
E[X] = n p, \qquad Var(X) = n p (1 - p)
$$

---

- **Expected Value:** Average number of successes in $n$ trials  
- **Variance:** Degree of variability (uncertainty) around the mean

**Reading the formula:**  
> “If I repeat the experiment many times, the average number of successes will approach $np$,  
> and the typical deviation from this mean is $\sqrt{np(1-p)}$.”

---

## 5️⃣ Worked Example

**Scenario:**  
Flip a fair coin 5 times. Let $X =$ number of heads. Find $P(X = 3)$.

---

### Step 1: Identify Parameters

$$
n = 5, \quad p = 0.5, \quad x = 3
$$

---

### Step 2: Apply the PMF

$$
P(X = 3) = \binom{5}{3} (0.5)^3 (0.5)^2 = 10 \times 0.125 \times 0.25 = 0.3125
$$

**Interpretation:**  
There is a **31.25% chance** of getting exactly **3 heads** in 5 coin tosses.

This formula captures both:
- The **number of favorable arrangements** ($\binom{5}{3} = 10$)
- The **probability of each arrangement** ($p^3(1-p)^2 = 0.125 \times 0.25$)

---

### Step 3: Expected Value

$$
E[X] = n p = 5 \times 0.5 = 2.5
$$

---

### Step 4: Variance and Standard Deviation

$$
Var(X) = n p (1 - p) = 5 \times 0.5 \times 0.5 = 1.25
$$

$$
\sigma_X = \sqrt{Var(X)} = \sqrt{1.25} \approx 1.12
$$

---

###  Interpretation

- **Mean (2.5):** On average, 2.5 heads per 5 flips.  
- **Standard deviation (1.12):** The number of heads typically varies by about 1.1 from the mean.  
- **Symmetry:** Since $p = 0.5$, the distribution is symmetric around 2.5.

---

## 6️⃣ Advanced Interpretation

| Property | Condition | Behavior |
|-----------|------------|-----------|
| **Symmetry** | $p = 0.5$ | Distribution symmetric about the mean $np$ |
| **Right-skewed** | $p < 0.5$ | More probability mass on the right tail |
| **Left-skewed** | $p > 0.5$ | More probability mass on the left tail |

---

###  **Applications in Data Science**

- **Quality control:** Number of defective items in a batch  
- **Clinical trials:** Number of patients responding to treatment  
- **Survey analysis:** Number of respondents choosing a specific option  
- **Spam detection:** Number of spam messages in a sample of emails  

---

###  **Intuitive Reading of the PMF**

> “Think of it as first **counting all possible ways** an outcome can occur  
> and then **multiplying** by the **likelihood** of each sequence.  
> That gives the total probability of observing exactly $x$ successes.”

---

###  **Summary Table**

| Concept | Formula | Interpretation |
|----------|----------|----------------|
| Support | $x \in \{0, 1, 2, \dots, n\}$ | Possible number of successes |
| PMF | $f(x) = \binom{n}{x} p^x (1-p)^{n-x}$ | Probability of $x$ successes |
| Mean | $E[X] = np$ | Expected number of successes |
| Variance | $Var(X) = np(1-p)$ | Spread of success counts |
| Example | $n=5$, $p=0.5$, $E[X]=2.5$, $Var(X)=1.25$ | Coin flip illustration |

---

###  **Key Takeaway**

The **binomial distribution** quantifies the randomness of success counts in repeated, independent experiments.  
It serves as a **bridge** between **discrete probability** and **statistical inference**, forming the foundation for:

- **Confidence intervals for proportions**  
- **Hypothesis testing**  
- **Bayesian updating**  

and many **data science applications** where outcomes are binary.

---


```python
# ================================================================
# 4.2 Binomial Distribution
# ================================================================

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import binom

# ----------------------------------------------------
# Step 1: Define Parameters of the Binomial Distribution
# ----------------------------------------------------
n = 5       # number of trials
p = 0.5     # probability of success per trial
x = np.arange(0, n+1)

# ----------------------------------------------------
# Step 2: Compute the PMF and CDF
# ----------------------------------------------------
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

# ----------------------------------------------------
# Step 3: Compute Theoretical Mean and Variance
# ----------------------------------------------------
E_X = n * p
Var_X = n * p * (1 - p)
Std_X = np.sqrt(Var_X)

# ----------------------------------------------------
# Step 4: Monte Carlo Simulation
# ----------------------------------------------------
n_sim = 100000
samples = np.random.binomial(n, p, size=n_sim)
emp_mean = np.mean(samples)
emp_var = np.var(samples)

# ----------------------------------------------------
# Step 5: Display Analytical and Empirical Results
# ----------------------------------------------------
print("=== BINOMIAL DISTRIBUTION ===")
print(f"Parameters: n={n}, p={p}")
print(f"Theoretical Mean (E[X])     = {E_X:.4f}")
print(f"Theoretical Variance Var(X) = {Var_X:.4f}")
print(f"Standard Deviation σ(X)     = {Std_X:.4f}")
print(f"Empirical Mean (Simulation) = {emp_mean:.4f}")
print(f"Empirical Variance          = {emp_var:.4f}")
print("--------------------------------------------------\n")

# Step 5.1: Show PMF/CDF table
df = pd.DataFrame({
    "x (Successes)": x,
    "P(X=x)": pmf.round(5),
    "F(x)=P(X≤x)": cdf.round(5)
})
display(df)

# ----------------------------------------------------
# Step 6: Create 3 Visuals per Row
# ----------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
fig.suptitle(f"Binomial Distribution (n={n}, p={p}) – PMF | CDF | Simulation", fontsize=14, fontweight='bold')

# --- Plot 1: PMF ---
axes[0].bar(x, pmf, color='royalblue', edgecolor='black')
axes[0].set_title("Probability Mass Function (PMF)", fontsize=12)
axes[0].set_xlabel("x (Number of Successes)")
axes[0].set_ylabel("P(X = x)")
axes[0].grid(axis='y', linestyle=':', linewidth=0.6)

# --- Plot 2: CDF ---
axes[1].step(x, cdf, where='mid', color='crimson', linewidth=2)
axes[1].set_title("Cumulative Distribution Function (CDF)", fontsize=12)
axes[1].set_xlabel("x (Number of Successes)")
axes[1].set_ylabel("F(x) = P(X ≤ x)")
axes[1].grid(axis='both', linestyle=':', linewidth=0.6)

# --- Plot 3: Monte Carlo Simulation ---
axes[2].hist(samples, bins=np.arange(-0.5, n+1.5, 1), density=True,
             alpha=0.6, color='skyblue', edgecolor='black', label='Empirical')
axes[2].plot(x, pmf, 'ro--', linewidth=2, label='Theoretical')
axes[2].set_title(f"Monte Carlo Simulation ({n_sim:,} trials)", fontsize=12)
axes[2].set_xlabel("x (Number of Successes)")
axes[2].set_ylabel("Probability")
axes[2].legend()
axes[2].grid(axis='y', linestyle=':', linewidth=0.6)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# ----------------------------------------------------
# Step 7: Interpretation
# ----------------------------------------------------
print("Interpretation:")
print("- The PMF shows discrete probabilities for 0–n successes.")
print("- The CDF illustrates cumulative probability up to each x.")
print("- The simulated histogram closely aligns with theoretical PMF, confirming convergence.")
print("- Expected value (E[X]=np) = 2.5, centered at the symmetric midpoint for p=0.5.")
print("- Variance (Var(X)=np(1-p)) = 1.25, quantifying the dispersion around the mean.")
print("- For p ≠ 0.5, the distribution becomes skewed: right-skewed if p < 0.5, left-skewed if p > 0.5.")
print("- Binomial models underpin experiments in quality control, A/B testing, and survey analysis.")
print("- Conceptually, each bar in the PMF is the combination of all sequences that yield x successes.")
```

#  4.2 Binomial Distribution – Example 4.2.1

---

##  **Problem**

Suppose **25% of products** in a manufacturing process are defective.  
Three items are selected at random and inspected, classified as **defective (D)** or **non-defective (N)**.  
Find the **probability distribution** of the number of defective items, denoted by $X$.

---

## 1️⃣ Step 1: Define the Random Variable

Let  
$$
X = \text{number of defective items in the sample of 3 items}
$$

Parameters:
$$
n = 3, \quad p = 0.25
$$
where:
- $n =$ number of trials (items inspected)
- $p =$ probability of a defective item

Possible values:
$$
x = 0, 1, 2, 3
$$

---

### Sample Space ($S$):

$$
S = \{ DDD, DDN, DND, NDD, NND, NDN, DNN, NNN \}
$$

There are $2^3 = 8$ **equally possible sequences** of $D$ and $N$.

---

## 2️⃣ Step 2: Apply the Binomial PMF

The **probability mass function** (PMF) for a Binomial distribution is:

$$
P(X = x) =
\begin{cases}
\displaystyle \binom{n}{x} p^x (1 - p)^{n - x}, & x = 0, 1, 2, 3 \\
0, & \text{otherwise}
\end{cases}
$$

Substitute $n = 3$ and $p = 0.25$:

$$
P(X = x) = \binom{3}{x} (0.25)^x (0.75)^{3 - x}
$$

---

###  **Interpretation of the Formula**

- $\displaystyle \binom{3}{x}$ — Number of ways to choose $x$ defective items from 3.  
- $(0.25)^x$ — Probability that those $x$ items are defective.  
- $(0.75)^{3-x}$ — Probability that the remaining items are non-defective.

---

## 3️⃣ Step 3: Compute Probabilities

| $x$ | Formula | $P(X=x)$ | Decimal |
|-----|----------|-----------|-----------|
| 0 | $\binom{3}{0}(0.25)^0(0.75)^3$ | $1 \times 1 \times 0.421875$ | **0.4219** |
| 1 | $\binom{3}{1}(0.25)^1(0.75)^2$ | $3 \times 0.25 \times 0.5625$ | **0.4219** |
| 2 | $\binom{3}{2}(0.25)^2(0.75)^1$ | $3 \times 0.0625 \times 0.75$ | **0.1406** |
| 3 | $\binom{3}{3}(0.25)^3(0.75)^0$ | $1 \times 0.015625 \times 1$ | **0.0156** |

---

###  **Step 4: Probability Distribution Table**

| $X$ | 0 | 1 | 2 | 3 |
|-----|---|---|---|---|
| **P(X)** | 0.4219 | 0.4219 | 0.1406 | 0.0156 |

---

### **Interpretation**

- There is a **42.2% chance** that **none** or **exactly one** item is defective.  
- The probability of **two** defective items is smaller (**14.1%**).  
- The chance of **all three** being defective is **very low (1.56%)**.

Since $p = 0.25 < 0.5$, the distribution is **right-skewed** —  
more probability mass lies at **lower numbers of defects**.

---

## 5️⃣ Step 5: Expected Value and Variance

Expected number of defective items:
$$
E[X] = n p = 3 \times 0.25 = 0.75
$$

Variance:
$$
Var(X) = n p (1 - p) = 3 \times 0.25 \times 0.75 = 0.5625
$$

Standard deviation:
$$
\sigma_X = \sqrt{Var(X)} = \sqrt{0.5625} \approx 0.75
$$

---

###  **Interpretation**

On average, we expect **0.75 defective items** per sample of 3.  
Most samples will contain **0 or 1 defective**, which aligns with the PMF.

---

## 6️⃣ Step 6: Reading the Formula in Context

$$
P(X = x) = \binom{3}{x} (0.25)^x (0.75)^{3 - x}
$$

| $x$ | Scenario | Combinatorial Term | Explanation |
|-----|-----------|-------------------|--------------|
| 0 | No defective items | $\binom{3}{0} = 1$ | All three non-defective |
| 1 | One defective | $\binom{3}{1} = 3$ | One defective, two non-defective |
| 2 | Two defective | $\binom{3}{2} = 3$ | Two defective, one non-defective |
| 3 | All defective | $\binom{3}{3} = 1$ | All three defective |

---


```python
# 4.2 Binomial Distribution
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 3
p = 0.25
x = np.arange(0, n + 1)

# 1. Theoretical PMF, CDF
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

# 2. Theoretical moments
mean_theoretical = n * p
var_theoretical = n * p * (1 - p)
std_theoretical = np.sqrt(var_theoretical)

# 3. Monte Carlo simulation
N = 20000
samples = np.random.binomial(n, p, size=N)
empirical_counts = np.bincount(samples, minlength=n+1) / N
empirical_mean = samples.mean()
empirical_var = samples.var()
empirical_std = samples.std()

# 4. Create DataFrame summary
summary = pd.DataFrame({
    "x": x,
    "P(X=x)": pmf,
    "CDF": cdf,
    "Empirical P(X=x)": empirical_counts
})
summary["Difference"] = summary["Empirical P(X=x)"] - summary["P(X=x)"]
print("=== Theoretical vs Empirical Comparison ===\n")
print(summary.round(4))
print(f"\nE[X] = {mean_theoretical:.4f},  Var(X) = {var_theoretical:.4f},  SD = {std_theoretical:.4f}")
print(f"Empirical E[X] = {empirical_mean:.4f},  Var = {empirical_var:.4f},  SD = {empirical_std:.4f}")

# 5. Running mean (Law of Large Numbers visualization)
running_mean = np.cumsum(samples) / np.arange(1, N + 1)

# === 6 Visualizations (3 per row) ===
plt.figure(figsize=(18, 10))

# 1️⃣ PMF
plt.subplot(2, 3, 1)
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title('Theoretical PMF: Binomial(3, 0.25)')
plt.xlabel('x (Number of Defective Items)')
plt.ylabel('P(X=x)')
for i, v in enumerate(pmf):
    plt.text(i, v + 0.01, f"{v:.3f}", ha='center')

# 2️⃣ CDF
plt.subplot(2, 3, 2)
plt.step(x, cdf, where='mid', color='darkblue')
plt.title('Theoretical CDF')
plt.xlabel('x')
plt.ylabel('P(X ≤ x)')
plt.grid(True, linestyle='--', alpha=0.6)

# 3️⃣ Empirical Histogram vs Theoretical PMF
plt.subplot(2, 3, 3)
plt.hist(samples, bins=np.arange(-0.5, n+1.5, 1), density=True, color='lightcoral', edgecolor='black', alpha=0.6, label='Empirical')
plt.plot(x, pmf, 'o-', color='blue', label='Theoretical PMF')
plt.title('Empirical vs Theoretical PMF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()

# 4️⃣ Running Mean Convergence
plt.subplot(2, 3, 4)
plt.plot(running_mean, color='green')
plt.axhline(mean_theoretical, color='red', linestyle='--', label='E[X]')
plt.title('Convergence of Running Mean (LLN)')
plt.xlabel('Number of Simulations')
plt.ylabel('Running Mean')
plt.legend()
plt.grid(True, alpha=0.6)

# 5️⃣ Comparison Bar Chart
plt.subplot(2, 3, 5)
width = 0.35
plt.bar(x - width/2, pmf, width, label='Theoretical', color='navy', alpha=0.7)
plt.bar(x + width/2, empirical_counts, width, label='Empirical', color='orange', alpha=0.7)
plt.title('Comparison: Theoretical vs Empirical P(X=x)')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()

# 6️⃣ Empirical vs Theoretical CDF
plt.subplot(2, 3, 6)
plt.step(x, cdf, where='mid', color='blue', label='Theoretical CDF')
empirical_cdf = np.cumsum(empirical_counts)
plt.step(x, empirical_cdf, where='mid', color='red', linestyle='--', label='Empirical CDF')
plt.title('Empirical vs Theoretical CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True, alpha=0.6)

plt.tight_layout()
plt.show()
```

# 4.3 Continuous Uniform Distribution

---

## 1️⃣ Continuous Random Variable

A **continuous random variable** $X$ is one that can take **any real value** within a specified interval — not just distinct points.

---

### 🔹 Infinite Possibilities

For instance:

- Height of a chair: $X = 45.1, 45.12, 45.123, \ldots$
- Between any two values, there are **infinitely many possible values**.

Hence, $X$ is **uncountable** in its support.

---

### 🔹 Probability of an Exact Value is Zero

$$
P(X = a) = 0
$$

A single point in a continuum has **no measurable width**, therefore it carries **no probability mass**.

---

### 🔹 Probability is Meaningful Over Intervals

$$
P(a \le X \le b) > 0
$$

Probabilities are **defined over ranges**, not isolated points.

**Reading Tip:**  
> Always think in terms of ranges, not points.  
> For continuous variables, “how wide is the interval?” determines the probability.

---

## 2️⃣ Probability Density Function (PDF)

The **probability density function** (PDF) $f(x)$ describes how probability is distributed over the support of $X$.

---

### 📘 Definition

$$
P(c \le X \le d) = \int_{c}^{d} f(x)\,dx
$$

where:

- $f(x) \ge 0$ for all $x$
- and  
  $$
  \int_{-\infty}^{\infty} f(x)\,dx = 1
  $$

**Reading Tip:**  
> Think of $f(x)$ as the *density* of probability at $x$.  
> The **area under the curve** between $c$ and $d$ gives the probability that $X$ falls within $[c, d]$.

---

## 3️⃣ Continuous Uniform Distribution

A **continuous uniform distribution** arises when $X$ is **equally likely** to take **any value** in an interval $[a, b]$.

---

### 🔸 Probability Density Function

$$
f(x) =
\begin{cases}
\frac{1}{b - a}, & a \le x \le b \\
0, & \text{otherwise}
\end{cases}
$$

---

### Step-by-Step Reading

- $f(x) = \frac{1}{b - a}$  
  → Constant density across $[a, b]$.  
  All values are **equally probable**.

- For $x$ outside $[a, b]$, $f(x) = 0$  
  → $X$ cannot take values outside its **support**.

---

### 🔸 Probability of an Interval $[c, d]$

$$
P(c \le X \le d) = \int_{c}^{d} \frac{1}{b - a}\,dx = \frac{d - c}{b - a}
$$

**Interpretation:**  
> Probability is simply the **fraction of the total interval length** that $[c, d]$ covers.

---

## 4️⃣ Key Feature – Intervals vs. Points

For continuous variables:

$$
P(a \le X \le b) = P(a < X \le b) = P(a < X < b) = P(a \le X < b)
$$

Why? Because $P(X = a) = 0$.

**Simplifying Feature:**  
The way you write the interval endpoints **does not affect the probability**.

**Reading Tip:**  
> Probabilities are **areas under the PDF**, never exact points.

---

## 5️⃣ Worked Example

###  Scenario

Rod lengths are **uniformly distributed** between 10 cm and 20 cm.

$$
X \sim \text{Uniform}(10, 20)
$$

The PDF is:

$$
f(x) = \frac{1}{20 - 10} = 0.1
$$

---

### Step 1: Probability of Interval $[12, 15]$

$$
P(12 \le X \le 15) = \frac{15 - 12}{20 - 10} = \frac{3}{10} = 0.3
$$

**Interpretation:**  
There is a **30% chance** that a randomly selected rod has a length between **12 cm and 15 cm**.

---

###  Step 2: Probability of Exact Value

$$
P(X = 13) = 0
$$

Even though 13 cm is within the interval, the **probability of hitting exactly 13 cm** is **zero**.

---

## 6️⃣ Visualization Tip

The PDF for the uniform distribution is a **rectangle** over $[a, b]$:

- **Height** = $\frac{1}{b - a}$  
- **Width** = $(b - a)$  
- **Area** = $1$ (total probability)

---

## 7️⃣ Moments of the Uniform Distribution

### Expected Value:
$$
E[X] = \frac{a + b}{2}
$$

### Variance:
$$
Var(X) = \frac{(b - a)^2}{12}
$$

---

###  Reading Tip

- **Area under curve = Probability**
- **Height of curve = Density**
- **Width of interval = Fraction of probability mass**

---

##  Conceptual Summary

| Property | Expression | Interpretation |
|:--|:--|:--|
| PDF | $f(x) = \frac{1}{b - a}$ for $a \le x \le b$ | Equal likelihood across $[a,b]$ |
| CDF | $F(x) = \frac{x - a}{b - a}$ for $a \le x \le b$ | Linear increase in cumulative probability |
| Mean | $E[X] = \frac{a + b}{2}$ | Center of the interval |
| Variance | $Var(X) = \frac{(b - a)^2}{12}$ | Dispersion of values |
| Probability of interval | $P(c \le X \le d) = \frac{d - c}{b - a}$ | Proportion of interval length |
| Probability of a point | $P(X=a)=0$ | Exact point has zero probability |

---

### Insight

The continuous uniform distribution is a **foundational case** for all continuous probability models because it represents **maximum entropy** given bounded support:

> “If all you know is that a variable lies within $[a, b]$, and nothing else,  
> the uniform distribution is the most unbiased model possible.”

This principle underlies Bayesian prior selection, Monte Carlo sampling, and random number generation in computational statistics.

---

###Takeaways

- $f(x)$ is constant across $[a,b]$ → all values are equally likely.  
- $P(X = a) = 0$, but $P(a \le X \le b) > 0$.  
- $E[X] = (a+b)/2$, $Var(X) = (b-a)^2/12$.  
- Probabilities correspond to **areas**, not counts.  
- The uniform distribution forms the **baseline model** for continuous probability and simulation.


```python
# 4.3 Continuous Uniform Distribution
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters
a, b = 10, 20
x = np.linspace(8, 22, 500)

# 1. Theoretical PDF and CDF
pdf = uniform.pdf(x, loc=a, scale=b-a)
cdf = uniform.cdf(x, loc=a, scale=b-a)

# 2. Analytical moments
mean_theoretical = (a + b) / 2
var_theoretical = (b - a)**2 / 12
std_theoretical = np.sqrt(var_theoretical)

# 3. Probability of interval [12,15]
P_12_15 = (15 - 12) / (b - a)

# 4. Monte Carlo Simulation
N = 20000
samples = uniform.rvs(loc=a, scale=b-a, size=N)
empirical_mean = np.mean(samples)
empirical_var = np.var(samples)
empirical_std = np.std(samples)

# 5. Running mean convergence
running_mean = np.cumsum(samples) / np.arange(1, N + 1)

# === 6 Visualizations (3 per row) ===
plt.figure(figsize=(18, 10))

# 1️⃣ PDF
plt.subplot(2, 3, 1)
plt.plot(x, pdf, color='navy')
plt.fill_between(x, pdf, color='skyblue', alpha=0.3)
plt.title("PDF of Uniform(10,20)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axvline(a, color='black', linestyle='--')
plt.axvline(b, color='black', linestyle='--')
plt.text(10.1, 0.09, "a=10", fontsize=10)
plt.text(19.5, 0.09, "b=20", fontsize=10)

# 2️⃣ CDF
plt.subplot(2, 3, 2)
plt.plot(x, cdf, color='darkgreen')
plt.title("CDF of Uniform(10,20)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True, linestyle='--', alpha=0.6)

# 3️⃣ Probability of [12,15]
plt.subplot(2, 3, 3)
plt.plot(x, pdf, color='navy')
plt.fill_between(x, pdf, where=(x>=12) & (x<=15), color='orange', alpha=0.5)
plt.title("P(12 ≤ X ≤ 15) = 0.3")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True, alpha=0.6)

# 4️⃣ Monte Carlo Histogram vs PDF
plt.subplot(2, 3, 4)
plt.hist(samples, bins=30, density=True, color='lightcoral', edgecolor='black', alpha=0.6, label='Empirical')
plt.plot(x, pdf, 'k--', linewidth=2, label='Theoretical PDF')
plt.title("Empirical Histogram vs Theoretical PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()

# 5️⃣ Running Mean Convergence
plt.subplot(2, 3, 5)
plt.plot(running_mean, color='blue')
plt.axhline(mean_theoretical, color='red', linestyle='--', label=f"E[X]={mean_theoretical}")
plt.title("Convergence of Sample Mean (Law of Large Numbers)")
plt.xlabel("Number of Samples")
plt.ylabel("Running Mean")
plt.legend()
plt.grid(True, alpha=0.6)

# 6️⃣ Cumulative Probability Visualization
plt.subplot(2, 3, 6)
plt.hist(samples, bins=30, density=True, cumulative=True, histtype='step', color='red', linewidth=2, label='Empirical CDF')
plt.plot(x, cdf, color='black', linewidth=2, label='Theoretical CDF')
plt.title("Empirical vs Theoretical CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True, alpha=0.6)

plt.tight_layout()
plt.show()

# === Statistical Summary ===
print("\n=== Uniform(10,20) Summary ===")
print(f"E[X] (theoretical): {mean_theoretical:.4f}")
print(f"Var(X) (theoretical): {var_theoretical:.4f}")
print(f"Std(X): {std_theoretical:.4f}")
print(f"P(12 ≤ X ≤ 15): {P_12_15:.4f}")
print("\nEmpirical Estimates from Simulation:")
print(f"Mean: {empirical_mean:.4f}, Var: {empirical_var:.4f}, Std: {empirical_std:.4f}")
```

# 4.3 Continuous Uniform Distribution – Example 4.2 (University Restaurant Buffet)

---

##  Scenario

The amount of salad students take is **uniformly distributed** between **5 ounces** and **15 ounces**.

Let the random variable be  
$$X = \text{amount of salad taken (ounces)}$$

Distribution:  
$$X \sim U(a=5,\; b=15)$$

We will:

- Determine the PDF;
- Compute $E[X]$ and the standard deviation $\sigma_X$;
- Find the probability a student **stays hungry**, i.e. $P(X < 9)$;
- Give a visual/interpretive reading.

---

## 1️⃣ Step 1: PDF (Probability Density Function)

For a continuous uniform random variable on $[a,b]$ the PDF is

$$
f(x)=
\begin{cases}
\dfrac{1}{b-a}, & a \le x \le b, \\[6pt]
0, & \text{otherwise.}
\end{cases}
$$

Substitute $a=5$, $b=15$:

$$
f(x)=
\begin{cases}
\dfrac{1}{15-5}=\dfrac{1}{10}=0.1, & 5 \le x \le 15, \\[6pt]
0, & \text{otherwise.}
\end{cases}
$$

**Interpretation:** the PDF is a rectangle of height $0.1$ across $[5,15]$. The PDF value itself is a density (not a probability); probabilities arise from areas under the rectangle.

---

## 2️⃣ Step 2: Expected Value and Standard Deviation

Standard formulas for $X \sim U(a,b)$:

- Mean:
  $$
  E[X]=\frac{a+b}{2}
  $$
- Variance:
  $$
  \operatorname{Var}(X)=\frac{(b-a)^2}{12}
  $$
- Standard deviation:
  $$
  \sigma_X=\sqrt{\operatorname{Var}(X)}=\frac{b-a}{\sqrt{12}}
  $$

Substitute $a=5$, $b=15$:

- Mean:
  $$
  E[X]=\frac{5+15}{2}=10 \quad\text{(ounces)}
  $$

- Variance:
  $$
  \operatorname{Var}(X)=\frac{(15-5)^2}{12}=\frac{100}{12}\approx 8.333\ \text{(ounces}^2\text{)}
  $$

- Standard deviation:
  $$
  \sigma_X=\sqrt{8.333\ldots}\approx 2.887\ \text{ounces}
  $$

**Reading:** On average students take 10 ounces. Typical deviation from the mean is about 2.89 ounces.

---

## 3️⃣ Step 3: Probability of Staying Hungry ($X<9$)

Use the CDF for the uniform distribution. For $a\le x\le b$,

$$
F(x) = P(X \le x) = \frac{x-a}{b-a}.
$$

Compute $P(X<9)$. For continuous distributions $P(X<9)=P(X\le 9)=F(9)$, so

$$
F(9)=\frac{9-5}{15-5}=\frac{4}{10}=0.4.
$$

**Interpretation:** There is a **40%** chance a student takes less than 9 ounces (i.e., remains hungry under the chosen threshold).

Geometric reading: area of rectangle from $x=5$ to $x=9$ is $0.1\times(9-5)=0.4$.

---

## 4️⃣ Step 4: Visual & Intuitive Interpretation

- **PDF shape:** rectangle on $[5,15]$ with height $0.1$.  
- **Area = probability:** area under the rectangle over any subinterval $[c,d]\subseteq[5,15]$ equals $(d-c)/(b-a)$.  
  Example: area on $[5,9]$ is $(9-5)/10 = 0.4$ (matches $P(X<9)$).
- **Point probabilities:** $P(X = x) = 0$ for any single $x$; probability is always an area.

---

## 5️⃣ Step 5: Summary Table

| Quantity | Formula | Value | Interpretation |
|---:|:---:|:---:|:---|
| PDF $f(x)$ | $1/(b-a)$ on $[a,b]$ | $0.1$ for $5\le x\le15$ | Uniform density (height of rectangle) |
| Mean $E[X]$ | $(a+b)/2$ | $10$ oz | Average salad amount |
| Variance $\operatorname{Var}(X)$ | $(b-a)^2/12$ | $100/12 \approx 8.333$ | Spread (oz$^2$) |
| Std. dev. $\sigma_X$ | $\sqrt{\operatorname{Var}(X)}$ | $\approx 2.887$ oz | Typical deviation from mean |
| $P(X<9)$ | $(9-5)/(15-5)$ | $0.4$ | 40% chance of taking <9 oz (staying hungry) |

---

## 6️⃣ Key Takeaway

The continuous uniform distribution provides a simple, **maximum-entropy** model on a bounded interval: when only bounds are known, the uniform law is the least informative (most impartial) choice. For the buffet:

- Modeling $X\sim U(5,15)$ implies **no prior reason** to prefer any amount between 5 and 15 ounces.  
- Probabilities reduce to **ratios of lengths**: $P(c\le X\le d)=(d-c)/(b-a)$.  
- This makes the uniform distribution a convenient baseline for **simulation**, **sensitivity checks**, and as a noninformative prior in Bayesian analyses when only bounds are credible.

---


```python
# 4.3 Continuous Uniform Distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters
a, b = 5, 15
x = np.linspace(a - 2, b + 2, 500)
pdf = uniform.pdf(x, loc=a, scale=b - a)
cdf = uniform.cdf(x, loc=a, scale=b - a)

# Step 1: PDF
f_x = 1 / (b - a)

# Step 2: Expected Value and Standard Deviation
mean = (a + b) / 2
variance = (b - a)**2 / 12
std_dev = np.sqrt(variance)

# Step 3: Probability of Staying Hungry (X < 9)
P_X_less_9 = (9 - a) / (b - a)

# Step 4: Visualization (3 visuals per row)
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# --- Visualization 1: PDF ---
axs[0].plot(x, pdf, color='blue', lw=2)
axs[0].fill_between(x, pdf, where=(x >= 5) & (x <= 9), color='lightblue', alpha=0.5)
axs[0].set_title("PDF: Uniform(5,15)")
axs[0].set_xlabel("x (ounces)")
axs[0].set_ylabel("Density f(x)")
axs[0].grid(alpha=0.3)

# --- Visualization 2: CDF ---
axs[1].plot(x, cdf, color='green', lw=2)
axs[1].axhline(P_X_less_9, color='red', linestyle='--')
axs[1].axvline(9, color='red', linestyle='--')
axs[1].set_title("CDF: Uniform(5,15)")
axs[1].set_xlabel("x (ounces)")
axs[1].set_ylabel("F(x) = P(X ≤ x)")
axs[1].grid(alpha=0.3)

# --- Visualization 3: Simulation Histogram ---
samples = np.random.uniform(a, b, 10000)
axs[2].hist(samples, bins=30, density=True, color='orange', edgecolor='black', alpha=0.6)
axs[2].axvline(9, color='red', linestyle='--', label='x=9')
axs[2].plot(x, pdf, color='blue', lw=2, label='Theoretical PDF')
axs[2].set_title("Simulated Samples vs Theoretical PDF")
axs[2].set_xlabel("x (ounces)")
axs[2].set_ylabel("Density")
axs[2].legend()
axs[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Step 5: Print Summary
print("=== Continuous Uniform Distribution: University Buffet Example ===")
print(f"PDF f(x) = 1/(b-a) = {f_x:.3f}")
print(f"Mean E[X] = (a+b)/2 = {mean:.3f}")
print(f"Variance Var(X) = (b-a)^2/12 = {variance:.3f}")
print(f"Std Dev σX = sqrt(Var(X)) = {std_dev:.3f}")
print(f"P(X < 9) = (9 - a)/(b - a) = {P_X_less_9:.3f}")
```

#  4.4 Normal Distribution

---

## 1️⃣ Intuitive Understanding

The **normal distribution** (a.k.a. Gaussian) is arguably the most important continuous distribution in statistics because many real-world quantities arise from the aggregation of many small, independent effects.

A typical empirical histogram (e.g., heights of 10,000 adult males) exhibits:

- **Symmetry** about the mean  
- A **bell-shaped** density curve  
- Most observations **clustered near the mean**  
- Few observations in the extreme tails

**Applications (typical examples):**

| Phenomenon | Why approximately normal |
|---|---|
| Heights of adults | Sum of many genetic and environmental factors |
| IQ scores | Aggregate of many cognitive sub-processes |
| Weights of newborns | Biological variability + many small influences |
| Measurement errors | Central limit behavior of many small error sources |

---

## 2️⃣ Mathematical Definition

A random variable $X$ is **normally distributed** with mean $\mu$ and variance $\sigma^2$ (standard deviation $\sigma$) if its probability density function (PDF) is

$$
f(x) \;=\; \frac{1}{\sigma\sqrt{2\pi}}\;\exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right), \qquad x\in\mathbb{R}.
$$

**Parameters:**

- $\mu$ — location (mean, center of the distribution)  
- $\sigma^2$ — variance (controls spread)  
- $\sigma$ — standard deviation ($\sigma=\sqrt{\sigma^2}$)  

**Reading the formula:**

- The **exponential** term $\exp\!\big(-\tfrac{(x-\mu)^2}{2\sigma^2}\big)$ enforces **symmetry** about $\mu$ and determines how probability density decays as $|x-\mu|$ increases.  
- The **scaling factor** $(\sigma\sqrt{2\pi})^{-1}$ normalizes the total area to 1.

> **Important:** $f(x)$ is a **density** (height); probabilities are **areas** under the curve over intervals.

---

## 3️⃣ Core Properties

- **Symmetry:** $f(\mu - t) = f(\mu + t)$ for all $t$.  
- **Mean = Median = Mode = $\mu$**: the center, most probable value.  
- **Unbounded support:** tails extend to $\pm\infty$ (but probabilities shrink rapidly).  
- **68–95–99.7 empirical rule:**  
  - $\approx 68\%$ of mass within $\mu\pm\sigma$  
  - $\approx 95\%$ within $\mu\pm2\sigma$  
  - $\approx 99.7\%$ within $\mu\pm3\sigma$  

These rules give quick approximations for intervals around the mean.

---

## 4️⃣ Probabilities and Areas (Geometric Reading)

For continuous $X$:

- $P(X=x)=0$ for any single $x$.  
- Probabilities refer to intervals:

$$
P(a\le X\le b)=\int_a^b f(x)\,dx.
$$

Common probability queries:

- $P(X\ge 20)$ — area to the right of 20.  
- $P(-5 < X < 9)$ — area between $-5$ and $9$.  
- Discrete-style approximation: $P(X=19)\approx P(18.5\le X < 19.5)$ if bin-width interpretation is useful.

**Reading tip:** Always interpret probabilities as **areas under the PDF**.

---

## 5️⃣ Standardization (Z-score)

To compute probabilities for any normal $X\sim N(\mu,\sigma^2)$, standardize:

$$
Z \;=\; \frac{X-\mu}{\sigma} \quad\Longrightarrow\quad Z \sim N(0,1).
$$

Then

$$
P(a\le X\le b) \;=\; P\!\left(\frac{a-\mu}{\sigma} \le Z \le \frac{b-\mu}{\sigma}\right),
$$

and probabilities can be read from the **standard normal** table or computed with software.

**Intuition:** standardization rescales and recenters the variable so universal Z-tables (or software) apply.

---

## 6️⃣ Visualization and Parameter Effects

- **Peak location:** at $x=\mu$.  
- **Spread control:** $\sigma$ — smaller $\sigma$ → sharper (narrow) peak; larger $\sigma$ → flatter, wider curve.  
- **Shaded area:** probability for an interval (visualized by shading the curve between two vertical lines).

**Practical example:** If $\mu=172$ cm and $\sigma=6$ cm, then

$$
P(166 \le X \le 178) \approx 0.68
$$

by the 68–95–99.7 rule (i.e., $\mu\pm\sigma$).

---

## 7️⃣ Why the Normal Distribution Matters in Data Science

- **Modeling natural variability:** Many measurements are approximately normal.  
- **Statistical inference:** Confidence intervals, $t$-tests (large-sample approximations), and linear regression rely on (approximate) normality assumptions.  
- **Central Limit Theorem (CLT):** Sums (or averages) of many independent, identically distributed variables (with finite variance) tend toward a normal distribution — providing a theoretical foundation for normal-based methods.  
- **Computational convenience:** closed-form PDF, tractable moments, and well-tabulated quantiles.

---

## 8️⃣ Summary Table

| Property | Description |
|---:|:---|
| Shape | Symmetric, bell-shaped |
| Parameters | Mean $\mu$, SD $\sigma$ |

| Point probability | $P(X=x)=0$ (use intervals) |
| Interval probability | Area under curve between endpoints |
| Standardization | $Z=(X-\mu)/\sigma \sim N(0,1)$ |
| Empirical rule | 68–95–99.7 for $\mu\pm\sigma,\mu\pm2\sigma,\mu\pm3\sigma$ |
| Applications | Heights, errors, CLT-based inference, many natural and measurement processes |

---

##  Insights

- The normal distribution is often an **approximation**: real data deviate (skewness, kurtosis, multi-modality). Diagnostics (QQ-plots, skewness/kurtosis checks) are essential before assuming normality for inference.  
- **Robust alternatives** (e.g., nonparametric methods, heavy-tailed models) matter when normality fails.  
- In **Bayesian** and **frequentist** frameworks, the normal distribution's conjugacy and analytical properties make it a workhorse, but one should always check model fit and residual behavior in applied work.

#  4.5 Standard Normal Distribution

---

## 1. Definition

The **standard normal distribution** is a special case of the normal distribution where:  

- Mean: $ \mu = 0 $  
- Standard deviation: $ \sigma = 1 $

It is denoted as:  
$$ Z \sim N(0, 1) $$

**Key idea:**  
Standardization allows us to convert any normal variable  
$$ X \sim N(\mu, \sigma^2) $$  
to the standard normal variable $Z$, simplifying probability calculations.

---

## 2. Z-Score Transformation

To standardize a normal variable:  
$$ Z = \frac{X - \mu}{\sigma} $$

Where:  
- $X$ = original normal variable  
- $\mu$ = mean of $X$  
- $\sigma$ = standard deviation of $X$  
- $Z$ = standard normal variable  

**Interpretation:**  
The $Z$-score tells how many standard deviations a value $X$ is from the mean.  

- $Z > 0$ → above mean  
- $Z < 0$ → below mean  

This enables use of a single **standard normal table** for all normal distributions.

---

## 3. Standard Normal Table (Z-Table)

The Z-table provides **cumulative probabilities**:

$$ P(Z \le z) $$

- Columns: first two digits of $z$  
- Rows: second decimal place  
- Cumulative from the far left of the curve ($-\infty$) to $z$

**Example:**  
$$ P(Z \le 0.54) = 0.7054 $$  
→ About **70.54%** of the area lies to the left of $Z = 0.54$.

---

## 4. Computing Probabilities

### a) Right-tail probability  
$$ P(Z > 0.6) = 1 - P(Z \le 0.6) = 1 - 0.7257 = 0.2743 $$  
**Interpretation:** ~27.43% of values lie above $0.6\sigma$ from the mean.

---

### b) Probability between two Z-scores  
$$
P(0.25 \le Z < 0.78) = P(Z \le 0.78) - P(Z \le 0.25)
$$  
$$
= 0.7823 - 0.5987 = 0.1836
$$  
**Interpretation:** 18.36% of values lie between 0.25 and 0.78 standard deviations above the mean.

---

### c) Using symmetry  
$$
P(Z \le -0.3) = 1 - P(Z \le 0.3)
$$  
$$
= 1 - 0.6179 = 0.3821
$$  

Symmetry allows positive Z-values for most lookups.

---

## 5. Finding Z for a Given Probability

**Example 1:**  
$$ P(Z < a) = 0.6950 $$  
Lookup: $Z \approx 0.51$  
→ 69.5% of values lie below **0.51σ above the mean**

**Example 2:**  
$$ P(Z > b) = 0.242 $$  
Convert to left-tail:  
$$ P(Z \le b) = 1 - 0.242 = 0.758 $$  
Lookup: $Z \approx 0.70$  
→ 24.2% of values lie above **0.7σ above the mean**

---

## 6. Why Standardization is Powerful

- **Single reference table:** All normal distributions can be standardized.  
- **Comparability:** Different datasets can be compared on the same Z-scale.  
- **Simplifies calculations:**  

$$
P(a \le X \le b) = P\left( \frac{a - \mu}{\sigma} \le Z \le \frac{b - \mu}{\sigma} \right)
$$

---

## 7. Intuitive Interpretation

| Z-Score | Interpretation |
|----------|----------------|
| $Z = 0$ | Mean |
| $Z = 1$ | 1 SD above mean |
| $Z = -1$ | 1 SD below mean |

- Area under curve → probability  
- Positive Z → right of mean  
- Negative Z → left of mean  

Z-scores make probability computations **efficient and universal**.

---

## 8. Summary Table

| Concept | Formula / Table | Interpretation |
|----------|------------------|----------------|
| Standardization | $Z = (X - \mu)/\sigma$ | Converts any normal to standard normal |
| Cumulative probability | $P(Z \le z)$ | Area under curve from $-\infty$ to $z$ |
| Right-tail probability | $P(Z > z) = 1 - P(Z \le z)$ | Area to the right |
| Between two Z-scores | $P(a \le Z < b) = P(Z \le b) - P(Z \le a)$ | Area between $a$ and $b$ |
| Negative Z | $P(Z \le -z) = 1 - P(Z \le z)$ | Use symmetry |
| Find Z for probability | Inverse lookup from table | Reverse from cumulative probability |

---

>  **Reading Tip:**  
> The standard normal distribution forms the foundation of all inferential statistics — enabling hypothesis testing, p-values, and confidence intervals.


```python
# ----------------------------------------------------------
# 4.5 STANDARD NORMAL DISTRIBUTION — POSTDOC-LEVEL DEMONSTRATION
# ----------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1️⃣ PARAMETERS ----------------------------------------------------------
mu, sigma = 0, 1  # Standard Normal: N(0,1)
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# 2️⃣ COMPUTE KEY PROBABILITIES -----------------------------------------
p_left_0_6 = norm.cdf(0.6)
p_right_0_6 = 1 - p_left_0_6
p_between = norm.cdf(0.78) - norm.cdf(0.25)
p_negative = 1 - norm.cdf(0.3)

# Print core probabilities
print("P(Z ≤ 0.6) =", round(p_left_0_6, 4))
print("P(Z > 0.6) =", round(p_right_0_6, 4))
print("P(0.25 ≤ Z < 0.78) =", round(p_between, 4))
print("P(Z ≤ -0.3) =", round(p_negative, 4))

# 3️⃣ VISUALIZATIONS -----------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ---- (1) Standard Normal PDF -----------------------------------------
axes[0].plot(x, pdf, color='blue', lw=2)
axes[0].fill_between(x, pdf, where=(x > 0.6), color='red', alpha=0.3)
axes[0].set_title('Standard Normal PDF (Right Tail P(Z>0.6))')
axes[0].set_xlabel('Z')
axes[0].set_ylabel('Density')
axes[0].grid(True)

# ---- (2) CDF Visualization -------------------------------------------
axes[1].plot(x, cdf, color='green', lw=2)
axes[1].axvline(0.6, color='red', linestyle='--', label='Z=0.6')
axes[1].text(0.7, 0.5, f"P(Z ≤ 0.6)={p_left_0_6:.3f}", color='black')
axes[1].set_title('Cumulative Distribution Function (CDF)')
axes[1].set_xlabel('Z')
axes[1].set_ylabel('P(Z ≤ z)')
axes[1].legend()
axes[1].grid(True)

# ---- (3) Z-score Symmetry Visualization ------------------------------
axes[2].plot(x, pdf, color='purple', lw=2)
axes[2].fill_between(x, pdf, where=(x < -0.3), color='orange', alpha=0.4, label='P(Z ≤ -0.3)')
axes[2].fill_between(x, pdf, where=(x > 0.3), color='cyan', alpha=0.4, label='P(Z ≥ 0.3)')
axes[2].axvline(-0.3, color='orange', linestyle='--')
axes[2].axvline(0.3, color='cyan', linestyle='--')
axes[2].set_title('Symmetry: P(Z ≤ -0.3) = 1 - P(Z ≤ 0.3)')
axes[2].set_xlabel('Z')
axes[2].set_ylabel('Density')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()

# 4️⃣ ADDITIONAL: Z-SCORE VISUAL INTERPRETATION -------------------------
z_values = [-2, -1, 0, 1, 2]
areas = [norm.cdf(z) for z in z_values]

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, pdf, color='darkblue', lw=2)
for z in z_values:
    ax.axvline(z, color='gray', linestyle='--')
    ax.text(z, 0.02, f"Z={z}", ha='center')

ax.fill_between(x, pdf, where=(x < 0), color='lightblue', alpha=0.3)
ax.set_title("Z-Score Interpretation — Regions and Symmetry")
ax.set_xlabel("Z")
ax.set_ylabel("Density")
ax.grid(True)
plt.show()
```

#  Example: Standard Normal Distribution Transformation

---

## Given

Let  
$$ X \sim N(20, 16) $$  

where:  
- Mean: $ \mu = 20 $  
- Variance: $ \sigma^2 = 16 $  
- Standard deviation: $ \sigma = 4 $

We transform $X$ to the **standard normal variable** $Z$ using:  
$$ Z = \frac{X - \mu}{\sigma} = \frac{X - 20}{4} $$

---

## (a) Compute \( P(X < 23) \)

**Step 1: Standardize**

$$ Z = \frac{23 - 20}{4} = 0.75 $$

**Step 2: Express in terms of Z**

$$ P(X < 23) = P(Z < 0.75) $$

---

## (b) Compute \( P(X > 14) \)

**Step 1: Standardize**

$$ Z = \frac{14 - 20}{4} = -1.5 $$

**Step 2: Express in terms of Z**

$$ P(X > 14) = P(Z > -1.5) $$

---

## (c) Compute \( P(16 < X < 24.8) \)

**Step 1: Standardize both bounds**

$$ Z_1 = \frac{16 - 20}{4} = -1 $$
$$ Z_2 = \frac{24.8 - 20}{4} = 1.2 $$

**Step 2: Express in terms of Z**

$$ P(16 < X < 24.8) = P(-1 < Z < 1.2) $$

---

##  Interpretation

| Probability | Standardized Form | Meaning |
|--------------|-------------------|----------|
| \( P(X < 23) \) | \( P(Z < 0.75) \) | Below 0.75 SD above mean |
| \( P(X > 14) \) | \( P(Z > -1.5) \) | Above 1.5 SD below mean |
| \( P(16 < X < 24.8) \) | \( P(-1 < Z < 1.2) \) | Between -1 and 1.2 SD around mean |

---

>  **Key Idea:**  
> Transforming $X$ to $Z$ allows us to compute probabilities using the **standard normal table**, making the process universal for any normal distribution.


```python
# ------------------------------------------------------------
# Example: Standard Normal Distribution Transformation
# ------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1️⃣ Parameters -------------------------------------------------------------
mu, sigma = 20, 4  # mean and standard deviation
x = np.linspace(8, 32, 1000)
pdf = norm.pdf(x, mu, sigma)

# 2️⃣ Standardization and Probability Computations ---------------------------
# (a) P(X < 23)
z_a = (23 - mu) / sigma
p_a = norm.cdf(z_a)

# (b) P(X > 14)
z_b = (14 - mu) / sigma
p_b = 1 - norm.cdf(z_b)

# (c) P(16 < X < 24.8)
z1, z2 = (16 - mu) / sigma, (24.8 - mu) / sigma
p_c = norm.cdf(z2) - norm.cdf(z1)

# Print results
print(f"(a) P(X < 23)  = P(Z < {z_a:.2f})  = {p_a:.4f}")
print(f"(b) P(X > 14)  = P(Z > {z_b:.2f})  = {p_b:.4f}")
print(f"(c) P(16 < X < 24.8) = P({z1:.2f} < Z < {z2:.2f}) = {p_c:.4f}")

# 3️⃣ VISUALIZATIONS (3 per row) ---------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ---- (a) P(X < 23) --------------------------------------------------------
axes[0].plot(x, pdf, lw=2, color='blue')
axes[0].fill_between(x, pdf, where=(x < 23), color='blue', alpha=0.3)
axes[0].axvline(23, color='red', linestyle='--')
axes[0].set_title(f"(a) P(X < 23) = {p_a:.3f}\nZ = {z_a:.2f}")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Density")
axes[0].grid(True)

# ---- (b) P(X > 14) --------------------------------------------------------
axes[1].plot(x, pdf, lw=2, color='green')
axes[1].fill_between(x, pdf, where=(x > 14), color='green', alpha=0.3)
axes[1].axvline(14, color='red', linestyle='--')
axes[1].set_title(f"(b) P(X > 14) = {p_b:.3f}\nZ = {z_b:.2f}")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Density")
axes[1].grid(True)

# ---- (c) P(16 < X < 24.8) ------------------------------------------------
axes[2].plot(x, pdf, lw=2, color='purple')
axes[2].fill_between(x, pdf, where=((x > 16) & (x < 24.8)), color='purple', alpha=0.3)
axes[2].axvline(16, color='red', linestyle='--')
axes[2].axvline(24.8, color='red', linestyle='--')
axes[2].set_title(f"(c) P(16 < X < 24.8) = {p_c:.3f}\nZ = [{z1:.2f}, {z2:.2f}]")
axes[2].set_xlabel("X")
axes[2].set_ylabel("Density")
axes[2].grid(True)

plt.tight_layout()
plt.show()

# 4️⃣ Additional: Z-scale transformation visualization ----------------------
z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z, 0, 1)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(z, pdf_z, color='darkblue', lw=2)
ax.fill_between(z, pdf_z, where=(z > z_b), color='lightgreen', alpha=0.4, label=f'P(Z>{z_b:.2f})')
ax.fill_between(z, pdf_z, where=(z < z_a), color='lightblue', alpha=0.4, label=f'P(Z<{z_a:.2f})')
ax.fill_between(z, pdf_z, where=((z > z1) & (z < z2)), color='violet', alpha=0.4, label=f'P({z1:.2f}<Z<{z2:.2f})')

ax.set_title("Standard Normal (Z) Representation of X ∼ N(20,16)")
ax.set_xlabel("Z")
ax.set_ylabel("Density")
ax.legend()
ax.grid(True)
plt.show()
```

#  Example: Population Proportion Using the Standard Normal Distribution

---

##  Context

IQ scores are **normally distributed** as:

$$ X \sim N(\mu = 100, \sigma^2 = 225), \quad \text{where } \sigma = 15 $$

We will compute probabilities for different IQ scenarios using **standardization** and the **Z-distribution**.

---

##  Scenario 1: IQ ≥ 124

### Step 1: Standardize

$$ Z = \frac{124 - 100}{15} = 1.6 $$

### Step 2: Express probability

$$ P(X \ge 124) = P(Z \ge 1.6) $$

### Step 3: Use Z-table

$$ P(Z \ge 1.6) = 1 - P(Z \le 1.6) = 1 - 0.9452 = 0.0548 $$

**Interpretation:**  
≈ **5.48%** of the population have IQ ≥ 124.

---

##  Scenario 2: IQ ≤ 124

$$ P(X \le 124) = P(Z \le 1.6) = 0.9452 $$

**Interpretation:**  
≈ **94.52%** of the population have IQ ≤ 124.

---

##  Scenario 3: IQ between 110 and 130

### Step 1: Standardize both values

$$ Z_1 = \frac{110 - 100}{15} = 0.667, \quad Z_2 = \frac{130 - 100}{15} = 2 $$

### Step 2: Express probability

$$ P(110 \le X \le 130) = P(0.667 \le Z \le 2) $$

### Step 3: Use Z-table

$$ P(Z \le 2) = 0.9772, \quad P(Z \le 0.667) \approx 0.7475 $$

$$ P(0.667 \le Z \le 2) = 0.9772 - 0.7475 = 0.2297 $$

**Interpretation:**  
≈ **22.97%** of the population have IQs between **110 and 130**.

---

##  Scenario 4: IQ below the mean (X < 100)

### Step 1: Standardize

$$ Z = \frac{100 - 100}{15} = 0 $$

### Step 2: Compute probability

$$ P(X < 100) = P(Z < 0) = 0.5 $$

**Interpretation:**  
Exactly **half the population** have IQ below the mean — due to the symmetry of the normal curve.

---

##  Key Takeaways

| Concept | Formula | Interpretation |
|----------|----------|----------------|
| **Z-score transformation** | $ Z = \frac{X - \mu}{\sigma} $ | Converts any normal variable to standard normal form |
| **Right-tail probability** | $ P(X \ge x) = 1 - P(Z \le z) $ | Area to the right of z |
| **Interval probability** | $ P(a \le X \le b) = P(Z \le z_b) - P(Z \le z_a) $ | Area between two Z-scores |
| **Symmetry** | $ P(X < \mu) = P(X > \mu) = 0.5 $ | Equal halves on both sides of the mean |

---

>  **Insight:**  
> Standardization transforms the problem of computing probabilities for any normal distribution into one of computing areas under the **standard normal curve**.  
> This universality is what makes the **Z-distribution** the cornerstone of statistical inference.


```python
# ------------------------------------------------------------
# Example: Population Proportion Using the Standard Normal Distribution
# ------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu, sigma = 100, 15  # IQ mean and standard deviation
x = np.linspace(40, 160, 1000)
pdf = norm.pdf(x, mu, sigma)

# Scenario 1: IQ ≥ 124
z1 = (124 - mu) / sigma
p1 = 1 - norm.cdf(z1)

# Scenario 2: IQ ≤ 124
p2 = norm.cdf(z1)

# Scenario 3: 110 ≤ X ≤ 130
z3a, z3b = (110 - mu) / sigma, (130 - mu) / sigma
p3 = norm.cdf(z3b) - norm.cdf(z3a)

# Scenario 4: X < 100
z4 = (100 - mu) / sigma
p4 = norm.cdf(z4)

# Print probabilities
print(f"Scenario 1: P(X ≥ 124) = P(Z ≥ {z1:.2f}) = {p1:.4f}")
print(f"Scenario 2: P(X ≤ 124) = P(Z ≤ {z1:.2f}) = {p2:.4f}")
print(f"Scenario 3: P(110 ≤ X ≤ 130) = P({z3a:.3f} ≤ Z ≤ {z3b:.3f}) = {p3:.4f}")
print(f"Scenario 4: P(X < 100) = P(Z < {z4:.2f}) = {p4:.4f}")

# ------------------------------------------------------------
# Visualization: 3 plots per row
# ------------------------------------------------------------
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Scenario 1: IQ ≥ 124
axes[0, 0].plot(x, pdf, color='blue', lw=2)
axes[0, 0].fill_between(x, pdf, where=(x >= 124), color='blue', alpha=0.3)
axes[0, 0].axvline(124, color='red', linestyle='--')
axes[0, 0].set_title(f"Scenario 1: P(X ≥ 124) = {p1:.3f}\nZ = {z1:.2f}")
axes[0, 0].set_xlabel("IQ")
axes[0, 0].set_ylabel("Density")
axes[0, 0].grid(True)

# Scenario 2: IQ ≤ 124
axes[0, 1].plot(x, pdf, color='green', lw=2)
axes[0, 1].fill_between(x, pdf, where=(x <= 124), color='green', alpha=0.3)
axes[0, 1].axvline(124, color='red', linestyle='--')
axes[0, 1].set_title(f"Scenario 2: P(X ≤ 124) = {p2:.3f}\nZ = {z1:.2f}")
axes[0, 1].set_xlabel("IQ")
axes[0, 1].set_ylabel("Density")
axes[0, 1].grid(True)

# Scenario 3: 110 ≤ X ≤ 130
axes[0, 2].plot(x, pdf, color='purple', lw=2)
axes[0, 2].fill_between(x, pdf, where=((x >= 110) & (x <= 130)), color='purple', alpha=0.3)
axes[0, 2].axvline(110, color='red', linestyle='--')
axes[0, 2].axvline(130, color='red', linestyle='--')
axes[0, 2].set_title(f"Scenario 3: P(110 ≤ X ≤ 130) = {p3:.3f}\nZ ∈ [{z3a:.2f}, {z3b:.2f}]")
axes[0, 2].set_xlabel("IQ")
axes[0, 2].set_ylabel("Density")
axes[0, 2].grid(True)

# Scenario 4: IQ < 100
axes[1, 0].plot(x, pdf, color='orange', lw=2)
axes[1, 0].fill_between(x, pdf, where=(x < 100), color='orange', alpha=0.3)
axes[1, 0].axvline(100, color='red', linestyle='--')
axes[1, 0].set_title(f"Scenario 4: P(X < 100) = {p4:.3f}\nZ = {z4:.2f}")
axes[1, 0].set_xlabel("IQ")
axes[1, 0].set_ylabel("Density")
axes[1, 0].grid(True)

# Standard Normal Representation
z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z, 0, 1)
axes[1, 1].plot(z, pdf_z, lw=2, color='black')
axes[1, 1].fill_between(z, pdf_z, where=(z >= z1), color='blue', alpha=0.3, label=f"Z ≥ {z1:.2f}")
axes[1, 1].fill_between(z, pdf_z, where=((z >= z3a) & (z <= z3b)), color='purple', alpha=0.3, label=f"{z3a:.2f} ≤ Z ≤ {z3b:.2f}")
axes[1, 1].fill_between(z, pdf_z, where=(z < z4), color='orange', alpha=0.3, label=f"Z < {z4:.2f}")
axes[1, 1].set_title("Standard Normal Representation (μ=0, σ=1)")
axes[1, 1].set_xlabel("Z-score")
axes[1, 1].set_ylabel("Density")
axes[1, 1].legend()
axes[1, 1].grid(True)

# Empty subplot for balance
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

#  Reverse Engineering with the Standard Normal Distribution

---

##  Problem Statement

Let  

$$ X \sim N(4, 0.25) $$

Find the value of **x** such that:

$$ P(X < x) = 0.67 $$

---

##  Step 1: Identify the Corresponding Z-Score

From the **standard normal table**:

$$ P(Z < z) = 0.67 \Rightarrow z = 0.44 $$

---

##  Step 2: Rearrange the Standardization Formula

Recall the standardization relationship:

$$ Z = \frac{x - \mu}{\sigma} \Rightarrow x = \mu + z \cdot \sigma $$

Given:

$$ \mu = 4, \quad \sigma^2 = 0.25 \Rightarrow \sigma = 0.5 $$

Substitute values:

$$ x = 4 + 0.44(0.5) = 4.22 $$

---

##  **Conclusion**

The value of **x** such that  

$$ P(X < x) = 0.67 $$  

is:

$$ \boxed{x = 4.22} $$

---

### Interpretation

- 67% of the distribution lies **below 4.22**.  
- This process of finding **x from a given probability** is known as **inverse standardization** or **reverse lookup**.  
- It is commonly used to determine **percentiles**, **cutoff scores**, and **confidence bounds**.

---

>  **Insight:**  
> Reverse engineering via Z-scores leverages the cumulative distribution function (CDF).  
> The transformation  
> $$ x = \mu + z\sigma $$  
> is the **inverse mapping** from the standardized domain back to the original scale, fundamental in computing quantiles and thresholds in statistical modeling.


```python
# ------------------------------------------------------------
# Example: Reverse Engineering with the Standard Normal Distribution
# ------------------------------------------------------------
# X ~ N(μ = 4, σ² = 0.25)
# Find x such that P(X < x) = 0.67
# ------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu, sigma = 4, 0.5   # mean and std deviation
p = 0.67             # target cumulative probability

# Step 1: Find corresponding z-value
z = norm.ppf(p)  # inverse CDF for standard normal
# Step 2: Compute x from z
x_val = mu + z * sigma

# Display results
print(f"P(X < x) = {p}")
print(f"Z = {z:.2f}")
print(f"x = {x_val:.2f}")

# ------------------------------------------------------------
# Visualization setup (3 plots per row)
# ------------------------------------------------------------

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Normal PDF with shaded area for P(X < x)
axes[0].plot(x, pdf, lw=2, color='blue')
axes[0].fill_between(x, pdf, where=(x <= x_val), color='blue', alpha=0.3)
axes[0].axvline(x_val, color='red', linestyle='--', label=f"x = {x_val:.2f}")
axes[0].set_title(f"PDF: P(X < {x_val:.2f}) = {p}")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Density")
axes[0].legend()
axes[0].grid(True)

# Plot 2: Standard Normal equivalent (Z-domain)
z_vals = np.linspace(-3, 3, 1000)
pdf_z = norm.pdf(z_vals)
axes[1].plot(z_vals, pdf_z, lw=2, color='green')
axes[1].fill_between(z_vals, pdf_z, where=(z_vals <= z), color='green', alpha=0.3)
axes[1].axvline(z, color='red', linestyle='--', label=f"Z = {z:.2f}")
axes[1].set_title(f"Standard Normal: P(Z < {z:.2f}) = {p}")
axes[1].set_xlabel("Z-score")
axes[1].set_ylabel("Density")
axes[1].legend()
axes[1].grid(True)

# Plot 3: CDF representation (direct cumulative probability)
axes[2].plot(x, cdf, lw=2, color='purple')
axes[2].axvline(x_val, color='red', linestyle='--')
axes[2].axhline(p, color='gray', linestyle=':')
axes[2].scatter(x_val, p, color='red', zorder=5)
axes[2].set_title("CDF Representation")
axes[2].set_xlabel("X")
axes[2].set_ylabel("Cumulative Probability")
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

#  Reverse Engineering a Pass Mark Using the Standard Normal Distribution

---

##  Scenario

Examination scores follow a normal distribution:

$$ \mu = 62, \quad \sigma = 16 $$

Thus,

$$ X \sim N(62, 256) $$

**Objective:**  
Determine the **minimum pass mark** such that **86% of candidates pass**.

---

##  Step 1: Translate the Probability

We are told:

$$ P(X > x) = 0.86 $$

This means only 14% of candidates score **below** the pass mark:

$$ P(X < x) = 0.14 $$

---

##  Step 2: Find the Corresponding Z-Score

From the **standard normal distribution table**:

$$ P(Z < z) = 0.14 \Rightarrow z = -1.08 $$

---

##  Step 3: Solve for \( x \)

Recall the standardization formula:

$$ Z = \frac{x - \mu}{\sigma} \Rightarrow x = \mu + z \cdot \sigma $$

Substitute known values:

$$
\begin{align*}
x &= 62 + (-1.08)(16) \\
x &= 62 - 17.28 \\
x &= 44.72
\end{align*}
$$

---

## **Conclusion**

To ensure that **86% of candidates pass**, the **minimum pass mark** should be set at approximately:

$$ \boxed{x = 44} $$

---

## Interpretation

- 14% of students score below **44**, while 86% score **above** it.
- This cutoff represents the **14th percentile** of the score distribution.
- By adjusting the pass mark, examiners control the **passing proportion**.

---

>  **Insight:**  
> This is an example of **quantile estimation** using the **inverse CDF (percent-point function)** of the normal distribution.  
> Setting thresholds based on cumulative probabilities is a core method in:
> - Standardized testing calibration,  
> - Quality control (defect rate thresholds), and  
> - Risk modeling (VaR computation).  
>
> In general form:
> $$ x_p = \mu + z_p \sigma $$
> where \( z_p \) is the quantile corresponding to \( P(Z < z_p) = p \).


```python
# ------------------------------------------------------------
# Example: Reverse Engineering a Pass Mark Using Standard Normal Distribution
# ------------------------------------------------------------
# Given: X ~ N(μ = 62, σ = 16)
# Find x such that P(X > x) = 0.86  =>  P(X < x) = 0.14
# ------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu, sigma = 62, 16
p = 0.14  # Left-tail probability (14% below cutoff)

# Step 1: Find corresponding Z-score
z = norm.ppf(p)  # Inverse CDF (percent point function)
# Step 2: Compute pass mark
x_val = mu + z * sigma

# Output results
print(f"P(X > x) = 0.86 => P(X < x) = {p}")
print(f"Z = {z:.2f}")
print(f"Pass mark (x) = {x_val:.2f}")

# ------------------------------------------------------------
# Visualization setup — 3 plots per row
# ------------------------------------------------------------

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: PDF with shaded region for P(X < x)
axes[0].plot(x, pdf, lw=2, color='blue')
axes[0].fill_between(x, pdf, where=(x <= x_val), color='blue', alpha=0.3)
axes[0].axvline(x_val, color='red', linestyle='--', label=f"x = {x_val:.2f}")
axes[0].set_title(f"PDF: P(X < {x_val:.2f}) = {p}")
axes[0].set_xlabel("Exam Score (X)")
axes[0].set_ylabel("Density")
axes[0].legend()
axes[0].grid(True)

# Plot 2: Standard Normal Representation
z_vals = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z_vals)
axes[1].plot(z_vals, pdf_z, lw=2, color='green')
axes[1].fill_between(z_vals, pdf_z, where=(z_vals <= z), color='green', alpha=0.3)
axes[1].axvline(z, color='red', linestyle='--', label=f"Z = {z:.2f}")
axes[1].set_title(f"Standard Normal: P(Z < {z:.2f}) = {p}")
axes[1].set_xlabel("Z-Score")
axes[1].set_ylabel("Density")
axes[1].legend()
axes[1].grid(True)

# Plot 3: CDF with cutoff point
axes[2].plot(x, cdf, lw=2, color='purple')
axes[2].axvline(x_val, color='red', linestyle='--')
axes[2].axhline(p, color='gray', linestyle=':')
axes[2].scatter(x_val, p, color='red', zorder=5)
axes[2].set_title("CDF Representation (Inverse Lookup)")
axes[2].set_xlabel("Exam Score (X)")
axes[2].set_ylabel("Cumulative Probability")
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

#  Example 1: Probability of Success in IT Startups

---

##  Problem Statement

80% of IT business startups generate a profit in their first year.  
A sample of 10 startups is selected.  
Determine the probability that **exactly 7 startups** will generate a profit.

---

##  Step 1: Verify Binomial Conditions

To apply the **binomial distribution**, check that:

- **Two mutually exclusive outcomes:**
  - Success ($S$) → generates profit  
  - Failure ($F$) → does not generate profit  

- **Fixed number of trials:**  
  $n = 10$ startups  

- **Independence of trials:**  
  Reasonable to assume each startup’s outcome does not affect others  

- **Constant probability of success:**  
  $p = 0.8$, hence $q = 1 - p = 0.2$

All conditions for a binomial model are satisfied.

---

##  Step 2: Define the Random Variable

Let $X$ = number of startups that generate a profit.  

Then:

$$ X \sim \text{Binomial}(n = 10, p = 0.8) $$

We are asked to find:

$$ P(X = 7) $$

---

##  Step 3: Apply the Binomial Formula

The **binomial probability mass function (PMF)** is:

$$ P(X = x) = \binom{n}{x} p^x q^{n - x} $$

Substitute values:

$$ P(X = 7) = \binom{10}{7} (0.8)^7 (0.2)^3 $$

---

##  Step 4: Compute the Probability

Compute the binomial coefficient:

$$ \binom{10}{7} = \frac{10!}{7!(10 - 7)!} = 120 $$

Compute powers:

$$ (0.8)^7 \approx 0.2097, \quad (0.2)^3 = 0.008 $$

Now multiply all components:

$$ P(X = 7) = 120 \times 0.2097 \times 0.008 \approx 0.2013 $$

---

##  Step 5: Interpretation

$$ P(X = 7) \approx 0.2013 $$

There is a **20.13% probability** that exactly 7 out of 10 startups will generate a profit in their first year.

This is consistent with expectations, since the probability of success $p = 0.8$ is high; therefore, values around $n \cdot p = 8$ are most likely, while 7 is slightly below the expected number.

---

##  Step 6: Practical Notes

Binomial probabilities can also be computed efficiently using statistical tools (e.g., Excel, R, or Python):

**Excel formula:**

BINOM.DIST(x, n, p, FALSE)

- Setting the last argument to `FALSE` returns the exact probability $P(X = x)$  
- Setting it to `TRUE` returns the cumulative probability $P(X \le x)$  

---

##  Step 7: Insights for Data Science

- **Binary outcome modeling:**  
  Binomial distribution models success/failure events (e.g., startup success, customer conversion, defect detection).

- **Expected number of successes:**

  $$ E[X] = n \cdot p = 10 \times 0.8 = 8 $$

- **Variance and spread:**

  $$ \text{Var}(X) = n p q = 10 \times 0.8 \times 0.2 = 1.6 $$

  $$ \sigma_X = \sqrt{1.6} \approx 1.26 $$

Thus, most likely outcomes are clustered around 7–9 successes.

---

## Takeaway

This example illustrates **discrete probability modeling** for real-world business analytics.

Key insights:
- Verify assumptions (independence, fixed $n$, constant $p$)
- Connect theoretical expectations ($E[X]$, $\text{Var}(X)$) to observed outcomes
- Understand how **probability mass functions** inform decision-making under uncertainty

Mastering such analysis enables predictive modeling for success/failure systems in **data science and business forecasting**.


```python
# ------------------------------------------------------------
# Example 1: Probability of Success in IT Startups
# ------------------------------------------------------------
# Given:
#   n = 10 startups
#   p = 0.8 (probability of success)
#   q = 0.2 (probability of failure)
# Task: Compute P(X = 7) for X ~ Binomial(10, 0.8)
# ------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10
p = 0.8
q = 1 - p
x = 7

# Step 1: Compute exact probability
P_X_7 = binom.pmf(x, n, p)

# Step 2: Expected value and variance
mean = n * p
var = n * p * q
std = np.sqrt(var)

# Display computed results
print("=== Binomial Distribution: IT Startups Example ===")
print(f"P(X = 7) = {P_X_7:.4f}")
print(f"Expected value E[X] = {mean:.2f}")
print(f"Variance Var(X) = {var:.2f}")
print(f"Std deviation σX = {std:.2f}")

# ------------------------------------------------------------
# Step 3: Visualization (3 plots per row)
# ------------------------------------------------------------

x_vals = np.arange(0, n + 1)
pmf_vals = binom.pmf(x_vals, n, p)
cdf_vals = binom.cdf(x_vals, n, p)
samples = binom.rvs(n, p, size=10000)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- Plot 1: PMF (Probability Mass Function) ---
axes[0].bar(x_vals, pmf_vals, color='skyblue', edgecolor='black')
axes[0].bar(x, pmf_vals[x], color='orange', edgecolor='black', label=f"P(X=7)={P_X_7:.2f}")
axes[0].set_title("PMF: Binomial(10, 0.8)")
axes[0].set_xlabel("Number of Profitable Startups (X)")
axes[0].set_ylabel("P(X = x)")
axes[0].legend()
axes[0].grid(alpha=0.3)

# --- Plot 2: CDF (Cumulative Distribution Function) ---
axes[1].step(x_vals, cdf_vals, where='mid', color='green', lw=2)
axes[1].axhline(y=binom.cdf(7, n, p), color='red', linestyle='--')
axes[1].axvline(x, color='red', linestyle='--', label=f"CDF at X=7")
axes[1].set_title("CDF: Cumulative Probability up to X")
axes[1].set_xlabel("X (Number of Profitable Startups)")
axes[1].set_ylabel("P(X ≤ x)")
axes[1].legend()
axes[1].grid(alpha=0.3)

# --- Plot 3: Simulated Histogram vs. Theoretical PMF ---
axes[2].hist(samples, bins=np.arange(-0.5, n + 1.5, 1), density=True, color='lightgray', edgecolor='black', alpha=0.7, label='Simulated')
axes[2].plot(x_vals, pmf_vals, 'o-', color='red', lw=2, label='Theoretical PMF')
axes[2].set_title("Simulation vs. Theoretical PMF")
axes[2].set_xlabel("Number of Profitable Startups (X)")
axes[2].set_ylabel("Relative Frequency / Probability")
axes[2].legend()
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# Step 4: Interpretation
# ------------------------------------------------------------
print("\nInterpretation:")
print(f"There is a {P_X_7*100:.2f}% probability that exactly 7 out of 10 startups will generate profit.")
print(f"Expected number of profitable startups: {mean:.1f}")
print("Distribution is centered near 8 successes, with outcomes 7–9 most likely, consistent with high success rate (p=0.8).")
```

#  Example 2: Probability of Selecting Hearts in Multiple Draws

---

##  Problem Statement

A card is randomly drawn from a standard deck of 52 cards **with replacement**, and this experiment is repeated **5 times**.  
Find the probability of selecting **exactly 3 hearts**.

---

##  Step 1: Verify Binomial Conditions

To apply the **binomial distribution**, we check:

- **Two mutually exclusive outcomes per trial:**
  - Success ($S$) → drawing a heart  
  - Failure ($F$) → drawing any other suit  

- **Fixed number of trials:**  
  $n = 5$ draws  

- **Independence of trials:**  
  True because cards are replaced after each draw  

- **Constant probability of success:**  
  $$ p = \frac{13}{52} = 0.25, \quad q = 1 - p = 0.75 $$

All conditions for a binomial model are satisfied.

---

##  Step 2: Define the Random Variable

Let $X$ = number of hearts drawn in 5 trials.  

Then:

$$ X \sim \text{Binomial}(n = 5, p = 0.25) $$

We are asked to find:

$$ P(X = 3) $$

---

##  Step 3: Apply the Binomial Formula

The **binomial probability mass function (PMF)** is:

$$ P(X = x) = \binom{n}{x} p^x q^{n - x} $$

Substitute the values:

$$ P(X = 3) = \binom{5}{3} (0.25)^3 (0.75)^2 $$

---

##  Step 4: Compute Each Component

Binomial coefficient:

$$ \binom{5}{3} = \frac{5!}{3!(5 - 3)!} = 10 $$

Powers:

$$ (0.25)^3 = 0.015625, \quad (0.75)^2 = 0.5625 $$

Multiply all components:

$$ P(X = 3) = 10 \times 0.015625 \times 0.5625 \approx 0.0879 $$

---

##  Step 5: Interpretation

$$ P(X = 3) \approx 0.0879 $$

There is an **8.79% probability** that exactly **3 out of 5 draws** will be hearts.  

Since $p = 0.25$, the most likely outcomes are 0 or 1 heart, making 3 hearts relatively unlikely.

---

##  Step 6: Statistical Insights

- **Expected value:**

  $$ E[X] = n \cdot p = 5 \times 0.25 = 1.25 $$

  On average, we expect about **1 heart** in 5 draws.

- **Variance and spread:**

  $$ \text{Var}(X) = n p q = 5 \times 0.25 \times 0.75 = 0.9375 $$

  $$ \sigma_X = \sqrt{0.9375} \approx 0.968 $$

  Most outcomes cluster around **1 heart**, with 2–3 hearts being increasingly rare.

- **Independence matters:**  
  Replacement ensures independence; without replacement, the probabilities would follow a **hypergeometric distribution** instead.

---

##  Step 7: Practical Note

Binomial probabilities like this can be quickly computed in statistical software (e.g., Excel, R, or Python).

**Excel Formula Example:**

BINOM.DIST(x = 3, n = 5, p = 0.25, FALSE)

- Returns $P(X = 3)$ when `FALSE`  
- Returns $P(X \le 3)$ when `TRUE`  

---

##  Takeaway

This example demonstrates **binomial probability modeling** in repeated random experiments.

Key insights:
- Verifying **independence and replacement** is essential for the binomial assumption.  
- **Expected values and variance** describe likely outcomes and variability.  
- The framework applies to numerous binary-event systems — from card games to reliability testing and quality control.

Understanding such distributions enhances probabilistic reasoning in **data science and stochastic modeling**.

#  Example 3: Probability of Winning 3 Games in a 5-Game Series

---

##  Problem Statement

A basketball team plays a series of **5 games** against an opponent.  
The winner of the series is the team that wins the **majority of games**.

Assume:

- Probability of winning a single game:  
  $$ p = 0.75 $$
- Probability of losing a single game:  
  $$ q = 1 - p = 0.25 $$

We want to find the probability that the team wins **exactly 3 games**.

---

##  Step 1: Verify Binomial Conditions

The **binomial distribution** applies because:

- Two possible outcomes per trial: **win or lose**  
- Fixed number of trials:  
  $$ n = 5 $$  
- Independent games: each game outcome is independent  
- Constant probability of success:  
  $$ p = 0.75 $$

All conditions for a binomial model are satisfied.

---

##  Step 2: Define the Random Variable

Let $X$ = number of wins in 5 games:

$$ X \sim \text{Binomial}(n = 5, p = 0.75) $$

We need:

$$ P(X = 3) $$

---

##  Step 3: Apply the Binomial Formula

The **binomial probability mass function (PMF)** is:

$$ P(X = x) = \binom{n}{x} p^x q^{n - x} $$

Substitute the values:

$$ P(X = 3) = \binom{5}{3} (0.75)^3 (0.25)^2 $$

---

##  Step 4: Compute Each Component

Binomial coefficient:

$$ \binom{5}{3} = \frac{5!}{3!(5 - 3)!} = 10 $$

Powers:

$$ (0.75)^3 = 0.421875, \quad (0.25)^2 = 0.0625 $$

Multiply all components:

$$ P(X = 3) = 10 \times 0.421875 \times 0.0625 = 0.263671875 $$

$$ \Rightarrow P(X = 3) \approx 0.264 $$

---

##  Step 5: Interpretation

$$ P(X = 3) \approx 0.264 $$

There is a **26.4% probability** that the team will win exactly **3 games**.

Even with a high individual game win probability ($p = 0.75$), winning **exactly** 3 of 5 games is less likely than winning 4 or 5, which are closer to the expected value.

---

##  Step 6: Expected Value and Variance

- **Expected wins:**

  $$ E[X] = n \cdot p = 5 \cdot 0.75 = 3.75 $$

  On average, the team is expected to win about **3–4 games**.

- **Variance:**

  $$ \text{Var}(X) = n p q = 5 \cdot 0.75 \cdot 0.25 = 0.9375 $$

- **Standard Deviation:**

  $$ \text{SD}(X) = \sqrt{0.9375} \approx 0.968 $$

Thus, most likely outcomes cluster around **3 to 5 wins**, with 3 being slightly below the mean.

---
##  Step 7: Practical Insights

- **Binomial modeling** effectively predicts outcomes in sports, risk modeling, or repeated independent events.  
- **Expected value** and **standard deviation** provide insight into likely outcomes and uncertainty.  
- **Cumulative probabilities**, like winning **at least 3 games**, can inform strategic or probabilistic decision-making.

---

## Postdoc-Level Takeaway

This example illustrates **discrete probabilistic modeling** in a competitive scenario using the **binomial distribution**.

Key concepts:
- The PMF gives the **exact probability** of each outcome.  
- The **expected value and variance** show where most probabilities lie.  
- Even with high $p$, the most probable outcome is not always the expected value—distributions reflect spread and chance.

Understanding these principles deepens one’s ability to model real-world stochastic systems in **data science, sports analytics, and uncertainty quantification**.


```python
# Example 3: Probability of Winning 3 Games in a 5-Game Series
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 5         # Number of games
p = 0.75      # Probability of winning a single game
x = 3         # Wins of interest

# Step 1: Compute the exact probability P(X = 3)
prob_x3 = binom.pmf(x, n, p)

# Step 2: Generate distribution values
x_values = np.arange(0, n + 1)
pmf_values = binom.pmf(x_values, n, p)
cdf_values = binom.cdf(x_values, n, p)

# Step 3: Compute descriptive statistics
mean = n * p
var = n * p * (1 - p)
std = np.sqrt(var)

# Step 4: Create 3 visualizations per row
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ---- Plot 1: PMF ----
axes[0].bar(x_values, pmf_values, color='lightgreen', edgecolor='black')
axes[0].bar(x, prob_x3, color='orange', edgecolor='black', label=f'P(X=3)={prob_x3:.4f}')
axes[0].set_title('Binomial PMF (n=5, p=0.75)', fontsize=14)
axes[0].set_xlabel('Number of Wins (X)')
axes[0].set_ylabel('P(X=x)')
axes[0].legend()

# ---- Plot 2: CDF ----
axes[1].step(x_values, cdf_values, where='mid', color='purple', label='CDF')
axes[1].axhline(y=binom.cdf(x, n, p), color='orange', linestyle='--', label=f'P(X≤3)={binom.cdf(x,n,p):.4f}')
axes[1].set_title('Cumulative Distribution Function (CDF)', fontsize=14)
axes[1].set_xlabel('X')
axes[1].set_ylabel('P(X ≤ x)')
axes[1].legend()

# ---- Plot 3: Annotated Summary ----
axes[2].axis('off')
summary_text = (
    f"Parameters:\n"
    f"n = {n} games\n"
    f"p = {p}\n\n"
    f"Computed Probability:\n"
    f"P(X=3) = {prob_x3:.4f}\n\n"
    f"Distribution Insights:\n"
    f"E[X] = {mean:.2f}\n"
    f"Var[X] = {var:.4f}\n"
    f"SD[X] = {std:.3f}\n\n"
    f"Interpretation:\n"
    f"• 26.4% chance of winning exactly 3 games.\n"
    f"• Expected wins ≈ 3.75 (between 3 and 4).\n"
    f"• Most probable outcomes: 3–5 wins.\n\n"
    f"Insight:\n"
    f"Even with high win probability, exact outcomes\n"
    f"depend on stochastic spread around E[X]."
)
axes[2].text(0.05, 0.95, summary_text, fontsize=12, verticalalignment='top', fontfamily='monospace')

plt.tight_layout()
plt.show()
```

#  Example 4: Probability of Acres Burned in Forest Fires

---

##  Problem Statement

The average number of acres burned by **forest and range fires** in a large New Mexico county is:

- Mean:  
  $$ \mu = 4300 \text{ acres} $$
- Standard deviation:  
  $$ \sigma = 750 \text{ acres} $$

Assume the distribution of the number of acres burned is **normal**.

We are asked to find the probability that between **2,500** and **4,200** acres will be burned in a given year.

Formally:

$$ P(2500 < X < 4200) \, ? $$

---

##  Step 1: Standardize the Values

Convert $X$ to the **standard normal variable** $Z$ using:

$$ Z = \frac{X - \mu}{\sigma} $$

---

### For the lower limit:

$$ X = 2500 $$
$$ Z_1 = \frac{2500 - 4300}{750} = \frac{-1800}{750} = -2.40 $$

### For the upper limit:

$$ X = 4200 $$
$$ Z_2 = \frac{4200 - 4300}{750} = \frac{-100}{750} = -0.1333 $$

Thus:

$$ P(2500 < X < 4200) = P(-2.40 < Z < -0.13) $$

---

##  Step 2: Use the Standard Normal Table

From $Z$-table:

- $$ P(Z < -0.13) = 0.4483 $$
- $$ P(Z < -2.40) = 0.0082 $$

---

##  Step 3: Compute the Probability

$$ P(-2.40 < Z < -0.13) = P(Z < -0.13) - P(Z < -2.40) $$

$$ P(-2.40 < Z < -0.13) = 0.4483 - 0.0082 = 0.4401 $$

---

##  Step 4: Interpretation

$$ \boxed{P(2500 < X < 4200) = 0.4401} $$

There is a **44.01% probability** that between **2,500 and 4,200 acres** will be burned in any given year.

---

##  Takeaway

This example demonstrates **probability estimation using the standard normal distribution**.  

Key insights:

- **Standardization** transforms any normal variable to a common $Z$-scale for probability lookup.  
- The **area under the curve** between two $Z$-values represents the probability of that interval.  
- The process highlights the link between **raw data (X)** and **standardized metrics (Z)** for inferential applications.

Mastery of this process enables accurate interpretation of real-world continuous data, such as environmental risk modeling, weather events, and insurance analytics.


```python
# Example 4: Probability of Acres Burned in Forest Fires
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 4300     # mean acres burned
sigma = 750   # standard deviation
x1, x2 = 2500, 4200  # interval bounds

# Step 1: Standardize
z1 = (x1 - mu) / sigma
z2 = (x2 - mu) / sigma

# Step 2: Compute probabilities
p_z2 = norm.cdf(z2)
p_z1 = norm.cdf(z1)
prob = p_z2 - p_z1

# Step 3: Generate X and normal PDF
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)

# Step 4: Visualization (3 per row)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ---- Plot 1: Normal PDF with shaded region ----
axes[0].plot(x, pdf, color='blue', lw=2)
axes[0].fill_between(x, pdf, 0, where=(x >= x1) & (x <= x2), color='skyblue', alpha=0.6)
axes[0].axvline(mu, color='red', linestyle='--', label=f'Mean = {mu}')
axes[0].set_title('Normal Distribution of Acres Burned', fontsize=14)
axes[0].set_xlabel('Acres Burned (X)')
axes[0].set_ylabel('Probability Density')
axes[0].legend()

# ---- Plot 2: Standard Normal Z-Scale ----
z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z)
axes[1].plot(z, pdf_z, color='purple', lw=2)
axes[1].fill_between(z, pdf_z, 0, where=(z >= z1) & (z <= z2), color='violet', alpha=0.6)
axes[1].axvline(0, color='black', linestyle='--')
axes[1].set_title('Standard Normal (Z) Distribution', fontsize=14)
axes[1].set_xlabel('Z-score')
axes[1].set_ylabel('Density')

# ---- Plot 3: Summary & Interpretation ----
axes[2].axis('off')
summary_text = (
    f"Parameters:\n"
    f"μ = {mu}\nσ = {sigma}\n\n"
    f"Bounds:\n"
    f"X₁ = {x1},  X₂ = {x2}\n"
    f"Z₁ = {z1:.2f},  Z₂ = {z2:.2f}\n\n"
    f"Probability:\n"
    f"P({x1} < X < {x2}) = {prob:.4f}\n\n"
    f"Interpretation:\n"
    f"≈ 44.01% chance that burned acres fall\n"
    f"between 2,500 and 4,200 in a given year.\n\n"
    f"Insight:\n"
    f"Standardization links raw data (X) to universal Z-scores,\n"
    f"enabling cross-domain comparison and inference.\n"
)
axes[2].text(0.05, 0.95, summary_text, fontsize=12, fontfamily='monospace', va='top')

plt.tight_layout()
plt.show()
```

#  Example 5: Finding the 38th Percentile of Burnt Acres

---

##  Given Data

The number of acres burned by forest fires follows a normal distribution:

$$ X \sim N(\mu = 4300, \sigma = 750) $$

We are asked to find the value of $X$ corresponding to the **38th percentile** — that is, the number of acres below which 38% of all years fall.

---

##  Step 1: Interpret the Percentile

The 38th percentile means:

$$ P(X \leq X_{0.38}) = 0.38 $$

We standardize to the standard normal variable $Z$:

$$ P(Z \leq z_{0.38}) = 0.38 $$

---

##  Step 2: Find the Corresponding Z-Score

From standard normal tables or inverse cumulative functions:

$$ z_{0.38} \approx -0.31 $$

Interpretation:

A $Z$-score of **−0.31** means the value lies **0.31 standard deviations below the mean**.

---

##  Step 3: Transform Z Back to X

We use the standardization formula:

$$ Z = \frac{X - \mu}{\sigma} \quad \Rightarrow \quad X = \mu + Z \cdot \sigma $$

Substitute the given values:

$$ X = 4300 + (-0.31)(750) $$
$$ X = 4300 - 232.5 = 4067.5 $$

---

##  Step 4: Interpretation

The **38th percentile** of burnt acres is:

$$ \boxed{X_{0.38} = 4067.5 \text{ acres}} $$

This means that **38% of the years** will have **fewer than 4068 acres burned**, and **62%** will have more.

---

##  Step 5: Key Insights

- Percentiles translate **probabilities** into **data values** within a normal distribution.  
- A **negative Z-score** indicates a value **below the mean**, while a positive one lies **above it**.  
- The transformation $X = \mu + Z\sigma$ allows us to **map between the Z-scale and the original data scale**.  
- Understanding percentiles supports **risk assessment, threshold setting, and policy planning** — essential in fields like environmental management, insurance, and data-driven decision-making.

---

 **Takeaway:**

Percentile analysis bridges **statistical inference** and **practical interpretation**.  
By combining the properties of the **standard normal distribution** with real-world data, we can quantify how typical or extreme an observation is — crucial for forecasting, risk modeling, and strategic decision-making.


```python
# Example 5: Finding the 38th Percentile of Burnt Acres
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mu = 4300      # mean
sigma = 750    # standard deviation
p = 0.38       # percentile (38%)

# Step 1: Find corresponding Z value
z_p = norm.ppf(p)   # inverse CDF for percentile
x_p = mu + z_p * sigma  # transform back to X scale

# Step 2: Generate distributions
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_x = norm.pdf(x, mu, sigma)

z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z)

# Step 3: Compute probabilities for shading
x_shade = x[x <= x_p]
pdf_shade = norm.pdf(x_shade, mu, sigma)

z_shade = z[z <= z_p]
pdf_z_shade = norm.pdf(z_shade)

# Step 4: Visualizations (3 plots per row)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- Plot 1: Normal distribution (X-scale) ---
axes[0].plot(x, pdf_x, color='navy', lw=2)
axes[0].fill_between(x_shade, pdf_shade, color='skyblue', alpha=0.6)
axes[0].axvline(mu, color='red', linestyle='--', label=f"Mean = {mu}")
axes[0].axvline(x_p, color='black', linestyle=':', lw=2, label=f"38th percentile = {x_p:.1f}")
axes[0].set_title("Normal Distribution of Acres Burned", fontsize=14)
axes[0].set_xlabel("Acres Burned (X)")
axes[0].set_ylabel("Probability Density")
axes[0].legend()

# --- Plot 2: Standard Normal (Z-scale) ---
axes[1].plot(z, pdf_z, color='purple', lw=2)
axes[1].fill_between(z_shade, pdf_z_shade, color='violet', alpha=0.6)
axes[1].axvline(0, color='red', linestyle='--', lw=1)
axes[1].axvline(z_p, color='black', linestyle=':', lw=2, label=f"z₀.₃₈ = {z_p:.2f}")
axes[1].set_title("Standard Normal Distribution", fontsize=14)
axes[1].set_xlabel("Z-score")
axes[1].set_ylabel("Density")
axes[1].legend()

# --- Plot 3: Summary & Postdoc Interpretation ---
axes[2].axis('off')
summary_text = (
    f"Percentile Computation Summary\n"
    f"{'-'*35}\n"
    f"Given: μ = {mu}, σ = {sigma}\n"
    f"Percentile: p = {p*100:.0f}%\n\n"
    f"Step 1: Find zₚ → z₀.₃₈ = {z_p:.2f}\n"
    f"Step 2: Transform to Xₚ → X = μ + zσ\n"
    f"= {mu} + ({z_p:.2f})×({sigma}) = {x_p:.1f}\n\n"
    f"Result:\nX₀.₃₈ = {x_p:.1f} acres\n\n"
    f"Interpretation:\n"
    f"38% of years burn ≤ {x_p:.1f} acres.\n"
    f"62% exceed this value.\n\n"
    f"Insight:\n"
    f"Percentiles quantify distribution thresholds.\n"
    f"They are essential in environmental risk assessment,\n"
    f"climate resilience modeling, and insurance analytics."
)
axes[2].text(0.05, 0.95, summary_text, fontsize=12, fontfamily='monospace', va='top')

plt.tight_layout()
plt.show()
```

#  Example: Percentage of Customers Spending Less Than \$3.00

---

##  Given Data

Concession spending follows a normal distribution:

$$ X \sim N(\mu = 4.11, \sigma = 1.37) $$

We want to find the percentage of customers who spend **less than \$3.00**.

---

##  Step 1: Standardize the Value

Convert the raw score $X = 3.00$ to a $Z$-score using:

$$ Z = \frac{X - \mu}{\sigma} $$

Substitute the values:

$$ Z = \frac{3.00 - 4.11}{1.37} = \frac{-1.11}{1.37} \approx -0.81 $$

 **Interpretation:**

A $Z$-score of **−0.81** means the spending amount is **0.81 standard deviations below the mean**.  
Negative $Z$-scores always correspond to values **below the average**.

---

##  Step 2: Find the Cumulative Probability

From standard normal distribution tables or statistical software:

$$ P(X < 3.00) = P(Z < -0.81) \approx 0.209 $$

That means about **20.9%** of the distribution lies below this value.

---

##  Step 3: Convert to Percentage

$$ 0.209 \times 100 = 20.9\% $$

**Interpretation:**

Approximately **21% of customers** spend **less than \$3.00** on concessions.  
This aligns with the mean of \$4.11, showing most customers spend more.

---

##  Step 4: Key Insights

- **Standardization** with $Z$-scores allows comparison across different scales and datasets.  
- **Percentiles** describe how a specific observation ranks within the entire population.  
- **Negative Z-scores** show how far below average a value is; **positive Z-scores** show how far above.  
- This analysis helps businesses with:
  - **Pricing strategy** (identifying low-spending segments)
  - **Inventory management**
  - **Revenue optimization**

---

 **Takeaway:**

Z-score standardization transforms any normal variable into a **universal probability scale**, enabling inference across diverse contexts.  
In applied analytics, such transformations support **customer segmentation**, **risk profiling**, and **strategic resource allocation** by quantifying how extreme or typical an observation is relative to the population.


```python
# Example: Percentage of Customers Spending Less Than $3.00
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mu = 4.11       # mean spending ($)
sigma = 1.37    # standard deviation ($)
x_value = 3.00  # spending threshold

# Step 1: Compute Z-score
z_value = (x_value - mu) / sigma

# Step 2: Compute cumulative probability
p = norm.cdf(z_value)
percentage = p * 100

# Step 3: Generate data for visualization
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_x = norm.pdf(x, mu, sigma)
x_shade = x[x <= x_value]
pdf_shade = norm.pdf(x_shade, mu, sigma)

z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z)
z_shade = z[z <= z_value]
pdf_z_shade = norm.pdf(z_shade)

# Step 4: Visualizations (3 per row)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- Plot 1: Normal distribution (raw spending) ---
axes[0].plot(x, pdf_x, color='navy', lw=2)
axes[0].fill_between(x_shade, pdf_shade, color='skyblue', alpha=0.6)
axes[0].axvline(mu, color='red', linestyle='--', label=f"Mean = ${mu}")
axes[0].axvline(x_value, color='black', linestyle=':', lw=2, label=f"x = ${x_value}")
axes[0].set_title("Customer Spending Distribution", fontsize=14)
axes[0].set_xlabel("Spending ($)")
axes[0].set_ylabel("Probability Density")
axes[0].legend()

# --- Plot 2: Standard Normal (Z-scale) ---
axes[1].plot(z, pdf_z, color='purple', lw=2)
axes[1].fill_between(z_shade, pdf_z_shade, color='violet', alpha=0.6)
axes[1].axvline(0, color='red', linestyle='--', lw=1)
axes[1].axvline(z_value, color='black', linestyle=':', lw=2, label=f"z = {z_value:.2f}")
axes[1].set_title("Standard Normal Transformation", fontsize=14)
axes[1].set_xlabel("Z-score")
axes[1].set_ylabel("Density")
axes[1].legend()

# --- Plot 3: Summary & Postdoc Interpretation ---
axes[2].axis('off')
summary_text = (
    f"Z-Score and Probability Analysis\n"
    f"{'-'*35}\n"
    f"Given: μ = {mu}, σ = {sigma}\n"
    f"Threshold: X = ${x_value}\n\n"
    f"Step 1: Z = (X - μ)/σ = ({x_value} - {mu})/{sigma} = {z_value:.2f}\n"
    f"Step 2: P(X < {x_value}) = P(Z < {z_value:.2f}) = {p:.3f}\n"
    f"Step 3: Convert → {p:.3f} × 100 = {percentage:.1f}%\n\n"
    f"Interpretation:\n"
    f"≈ {percentage:.1f}% of customers spend less than ${x_value}.\n"
    f"This spending lies {abs(z_value):.2f} SDs below the mean.\n\n"
    f"Insight:\n"
    f"Standardization enables comparison across populations.\n"
    f"Negative Z indicates below-average spending behavior.\n"
    f"Used in segmentation, risk modeling, and pricing analytics."
)
axes[2].text(0.05, 0.95, summary_text, fontsize=12, fontfamily='monospace', va='top')

plt.tight_layout()
plt.show()
```

#  Example: Probability of Scoring Above 500 on the GMAT

---

##  Given

GMAT scores follow a normal distribution:

$$ X \sim N(\mu = 527, \sigma = 112) $$

We are asked to find the probability that an individual scores **above 500**.

---

##  Step 1: Standardize the Value

Convert the raw score to a **Z-score**:

$$ Z = \frac{X - \mu}{\sigma} $$

Substitute the given values:

$$ Z = \frac{500 - 527}{112} = \frac{-27}{112} \approx -0.241 $$

**Interpretation:**

A Z-score of \( Z = -0.24 \) indicates the score is **0.24 standard deviations below the mean**.  
Negative Z-scores always correspond to values **below the mean**.

---

##  Step 2: Express the Probability in Terms of Z

We are looking for:

$$ P(X > 500) = P(Z > -0.24) $$

Using the **standard normal cumulative distribution function (CDF):**

$$ P(Z > -0.24) = 1 - P(Z \le -0.24) $$

From the **Z-table:**

$$ P(Z \le -0.24) \approx 0.4052 $$

Then:

$$ P(Z > -0.24) = 1 - 0.4052 = 0.5948 $$

---

##  Step 3: Interpretation

 **Result:**

$$ P(X > 500) \approx 0.5948 $$  
That is, **59.48% of test-takers** score above 500.

**Explanation:**

Since 500 is slightly **below the mean (527)**, it is expected that a **majority of individuals** score higher.

---

##  Step 4: Insights

- **Z-scores** provide a standardized way to compare scores across different scales.
- **Tail probabilities** (upper or lower) quantify proportions beyond thresholds — useful in ranking and decision-making.
- The **normal model** supports consistent interpretation across diverse datasets (e.g., admissions, performance, risk).

---

##  Takeaway

Through **standardization**, we can universally interpret performance metrics across domains.

This GMAT example illustrates the power of the **normal distribution** in quantifying relative standing, supporting data-driven insights in **educational assessment, selection thresholds, and benchmarking analysis**.


```python
# Example: Probability of Scoring Above 500 on the GMAT
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mu = 527     # mean GMAT score
sigma = 112  # standard deviation
x_value = 500

# Step 1: Compute Z-score
z_value = (x_value - mu) / sigma

# Step 2: Compute probability of scoring above 500
p_above = 1 - norm.cdf(z_value)

# Step 3: Generate data for visualization
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_x = norm.pdf(x, mu, sigma)
x_shade = x[x >= x_value]
pdf_shade = norm.pdf(x_shade, mu, sigma)

z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z)
z_shade = z[z >= z_value]
pdf_z_shade = norm.pdf(z_shade)

# Step 4: Visualization (3 plots)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- Plot 1: Normal Distribution of GMAT Scores ---
axes[0].plot(x, pdf_x, color='navy', lw=2)
axes[0].fill_between(x_shade, pdf_shade, color='skyblue', alpha=0.6)
axes[0].axvline(mu, color='red', linestyle='--', lw=1.5, label=f"Mean = {mu}")
axes[0].axvline(x_value, color='black', linestyle=':', lw=2, label=f"Score = {x_value}")
axes[0].set_title("GMAT Score Distribution", fontsize=14)
axes[0].set_xlabel("GMAT Score")
axes[0].set_ylabel("Probability Density")
axes[0].legend()

# --- Plot 2: Standard Normal (Z-scale) ---
axes[1].plot(z, pdf_z, color='purple', lw=2)
axes[1].fill_between(z_shade, pdf_z_shade, color='violet', alpha=0.6)
axes[1].axvline(0, color='red', linestyle='--', lw=1)
axes[1].axvline(z_value, color='black', linestyle=':', lw=2, label=f"z = {z_value:.2f}")
axes[1].set_title("Standard Normal Representation", fontsize=14)
axes[1].set_xlabel("Z-score")
axes[1].set_ylabel("Density")
axes[1].legend()

# --- Plot 3: Summary and Statistical Interpretation ---
axes[2].axis('off')
summary_text = (
    f"GMAT Upper-Tail Probability Analysis\n"
    f"{'-'*40}\n"
    f"Given: μ = {mu}, σ = {sigma}\n"
    f"Compute Z = (X - μ)/σ = ({x_value} - {mu})/{sigma} = {z_value:.2f}\n\n"
    f"Then P(X > 500) = 1 - P(Z ≤ {z_value:.2f})\n"
    f"               = 1 - {norm.cdf(z_value):.4f}\n"
    f"               = {p_above:.4f}\n\n"
    f"Interpretation:\n"
    f"≈ {p_above*100:.2f}% of test-takers score above 500.\n"
    f"Since 500 < mean(527), the probability exceeds 0.5.\n\n"
    f"Insight:\n"
    f"Upper-tail probabilities quantify relative performance.\n"
    f"Standardization enables fair cross-group comparisons,\n"
    f"essential for ranking, selection, and decision thresholds."
)
axes[2].text(0.05, 0.95, summary_text, fontsize=12, fontfamily='monospace', va='top')

plt.tight_layout()
plt.show()
```

#  Example: GMAT Score Threshold for Top 5% Performance

---

##  Given

GMAT scores are normally distributed as:

$$ X \sim N(\mu = 527, \sigma = 112) $$

We want to determine the **minimum score** required to rank in the **top 5%** of all test-takers.

---

##  Step 1: Translate the Probability to the Standard Normal

We are asked to find the score \( x \) such that:

$$ P(X > x) = 0.05 $$

Convert this to the standard normal form:

$$ P(Z > z) = 0.05 \implies P(Z < z) = 0.95 $$

From the **standard normal table** (or inverse cumulative distribution):

$$ z = 1.645 $$

**Interpretation:**

A Z-score of \( z = 1.645 \) corresponds to **1.645 standard deviations above the mean**, representing the **top 5%** of the distribution.

---

##  Step 2: Convert Z Back to the GMAT Score

Use the transformation formula:

$$ X = \mu + Z \cdot \sigma $$

Substitute the known values:

$$ X = 527 + (1.645)(112) $$

$$ X = 527 + 184.24 = 711.24 $$

**Minimum GMAT score for the top 5%:**

$$ X \approx 711 $$

---

##  Step 3: Interpretation

A score of **711 or higher** places a test-taker in the **top 5%** of GMAT performance.

This demonstrates how **percentiles** can be mapped to raw scores using the **mean and standard deviation** of a normal distribution.

Such transformations rely on **inverse Z-score lookups** (from tables or software).

---

##  Step 4: Insights

- High-percentile scores correspond to values **several standard deviations above the mean**.
- The **Z-score method** standardizes any normal variable for universal interpretation.
- Applications include:
  - **Admissions analytics**
  - **Performance benchmarking**
  - **Setting qualification thresholds**
- On a bell curve, \( X = 711 \) marks a **rare, elite tail region**, emphasizing exceptional performance.

---

##  Takeaway

This example illustrates **inverse normal inference**, where we determine the cutoff score for a target percentile.

By understanding the link between **Z-scores**, **percentiles**, and **distribution parameters**, analysts can design **data-driven thresholds** for selection, evaluation, and policy-making across diverse domains.


```python
# Example: GMAT Score Threshold for Top 5% Performance
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mu = 527     # mean GMAT score
sigma = 112  # standard deviation
p_top = 0.05 # top 5% cutoff

# Step 1: Find the Z corresponding to P(Z > z) = 0.05
z_crit = norm.ppf(1 - p_top)  # inverse CDF for 0.95 cumulative
x_crit = mu + z_crit * sigma  # convert back to GMAT scale

# Step 2: Generate data for visualization
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_x = norm.pdf(x, mu, sigma)
x_shade = x[x >= x_crit]
pdf_shade = norm.pdf(x_shade, mu, sigma)

z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z)
z_shade = z[z >= z_crit]
pdf_z_shade = norm.pdf(z_shade)

# Step 3: Visualization (3 plots per row)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- Plot 1: GMAT Score Distribution ---
axes[0].plot(x, pdf_x, color='navy', lw=2)
axes[0].fill_between(x_shade, pdf_shade, color='lightgreen', alpha=0.6)
axes[0].axvline(mu, color='red', linestyle='--', lw=1.5, label=f"Mean = {mu}")
axes[0].axvline(x_crit, color='black', linestyle=':', lw=2, label=f"Top 5% Threshold = {x_crit:.1f}")
axes[0].set_title("GMAT Score Distribution", fontsize=14)
axes[0].set_xlabel("GMAT Score")
axes[0].set_ylabel("Probability Density")
axes[0].legend()

# --- Plot 2: Standard Normal Representation ---
axes[1].plot(z, pdf_z, color='purple', lw=2)
axes[1].fill_between(z_shade, pdf_z_shade, color='violet', alpha=0.6)
axes[1].axvline(0, color='red', linestyle='--', lw=1)
axes[1].axvline(z_crit, color='black', linestyle=':', lw=2, label=f"z = {z_crit:.3f}")
axes[1].set_title("Standard Normal (Z) Domain", fontsize=14)
axes[1].set_xlabel("Z-score")
axes[1].set_ylabel("Density")
axes[1].legend()

# --- Plot 3: Analytical Summary ---
axes[2].axis('off')
summary_text = (
    f"GMAT Top 5% Threshold Analysis\n"
    f"{'-'*38}\n"
    f"Given: μ = {mu}, σ = {sigma}\n"
    f"Target: P(X > x) = 0.05\n"
    f"→ P(Z < z) = 0.95 → z = {z_crit:.3f}\n\n"
    f"x = μ + zσ = {mu} + ({z_crit:.3f})({sigma})\n"
    f"x = {x_crit:.2f}\n\n"
    f"Interpretation:\n"
    f"Only 5% of test-takers score above {x_crit:.0f}.\n"
    f"This value marks elite GMAT performance.\n\n"
    f"Insight:\n"
    f"Inverse normal inference (quantile mapping) underpins\n"
    f"admissions thresholds, performance benchmarking,\n"
    f"and tail-risk estimation in analytics contexts."
)
axes[2].text(0.05, 0.95, summary_text, fontsize=12, fontfamily='monospace', va='top')

plt.tight_layout()
plt.show()
```

#  Example: Proportion of Pregnancies Lasting Between 240 and 270 Days

---

##  Given

Pregnancy lengths are normally distributed as:

$$ X \sim N(\mu = 266, \sigma = 16) $$

We need to find the **proportion of pregnancies** lasting between **$240$** and **$270$** days.

##  Step 1: Standardize the Bounds

Convert both values to **$Z$-scores** using:

$$ Z = \frac{X - \mu}{\sigma} $$

$$
Z_1 = \frac{240 - 266}{16} = \frac{-26}{16} = -1.63
$$

$$
Z_2 = \frac{270 - 266}{16} = \frac{4}{16} = 0.25
$$

**Interpretation:**

- $Z_1 = -1.63$: $240$ days is **$1.63$ standard deviations below** the mean  
- $Z_2 = 0.25$: $270$ days is **$0.25$ standard deviations above** the mean  

---

## Step 2: Compute the Probability

We seek:

$$ P(240 < X < 270) = P(-1.63 < Z < 0.25) $$

Using the **standard normal cumulative probabilities**:

$$
P(Z < 0.25) = 0.5987
$$

$$
P(Z < -1.63) = 0.0516
$$

Then:

$$
P(-1.63 < Z < 0.25) = 0.5987 - 0.0516 = 0.5471
$$

**Result:**

$$ P(240 < X < 270) = 0.5471 $$

---

##  Step 3: Interpretation

Approximately **$54.71\%$** of pregnancies last **between $240$ and $270$ days**.

- This range corresponds roughly to **$8$–$9$ months**.  
- Slightly more than half of pregnancies fall in this interval.  
- The remaining **$45\%$** are shorter than $240$ days or longer than $270$ days.

---

##  Step 4: Insights

- **$Z$-score standardization** enables probability calculation across any normal distribution.  
- **Continuous probabilities** are represented by the **area under the bell curve**.  
- **Negative $Z$-scores** correspond to values **below** the mean, and **positive $Z$-scores** to values **above**.  
- Such analyses are useful in:
  - **Medical research**
  - **Public health policy**
  - **Predictive modeling of biological events**

---

## Takeaway

This example highlights the application of the **standard normal model** in real-world biological data.  
By mapping a normal variable to **$Z$-scores**, we quantify the proportion of outcomes within any range—  
a foundation for **risk modeling**, **clinical forecasting**, and **data-driven decision-making** in healthcare.


```python
# Example: Proportion of Pregnancies Lasting Between 240 and 270 Days
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -----------------------------
# GIVEN PARAMETERS
# -----------------------------
mu = 266     # mean pregnancy length (days)
sigma = 16   # standard deviation (days)
lower = 240  # lower bound
upper = 270  # upper bound

# -----------------------------
# STEP 1: STANDARDIZATION
# -----------------------------
z1 = (lower - mu) / sigma
z2 = (upper - mu) / sigma

# -----------------------------
# STEP 2: PROBABILITY CALCULATION
# -----------------------------
p1 = norm.cdf(z1)
p2 = norm.cdf(z2)
p_between = p2 - p1

# -----------------------------
# STEP 3: DATA FOR VISUALIZATION
# -----------------------------
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_x = norm.pdf(x, mu, sigma)

x_shade = x[(x >= lower) & (x <= upper)]
pdf_shade = norm.pdf(x_shade, mu, sigma)

z = np.linspace(-4, 4, 1000)
pdf_z = norm.pdf(z)

z_shade = z[(z >= z1) & (z <= z2)]
pdf_z_shade = norm.pdf(z_shade)

# -----------------------------
# STEP 4: VISUALIZATION
# -----------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Normal Distribution Analysis: Pregnancy Duration", fontsize=16, fontweight='bold')

# --- Plot 1: Original Distribution (X-scale) ---
axes[0].plot(x, pdf_x, color='navy', lw=2, label='Normal PDF')
axes[0].fill_between(x_shade, pdf_shade, color='lightgreen', alpha=0.7, label='P(240 < X < 270)')
axes[0].axvline(mu, color='red', linestyle='--', lw=1.5, label=f"Mean μ = {mu}")
axes[0].axvline(lower, color='black', linestyle=':', lw=2)
axes[0].axvline(upper, color='black', linestyle=':', lw=2)
axes[0].set_title("Pregnancy Length Distribution (X ~ N(266, 16²))", fontsize=13)
axes[0].set_xlabel("Pregnancy Duration (days)", fontsize=12)
axes[0].set_ylabel("Probability Density f(X)", fontsize=12)
axes[0].legend(fontsize=10, loc='upper left')
axes[0].grid(alpha=0.3)

# --- Plot 2: Standard Normal (Z-scale) ---
axes[1].plot(z, pdf_z, color='purple', lw=2, label='Standard Normal PDF')
axes[1].fill_between(z_shade, pdf_z_shade, color='violet', alpha=0.6, label=f"P({z1:.2f} < Z < {z2:.2f})")
axes[1].axvline(0, color='red', linestyle='--', lw=1.5, label="Mean (Z = 0)")
axes[1].axvline(z1, color='black', linestyle=':', lw=2)
axes[1].axvline(z2, color='black', linestyle=':', lw=2)
axes[1].set_title("Standard Normal Distribution (Z-Scores)", fontsize=13)
axes[1].set_xlabel("Z-Score (Standardized Units)", fontsize=12)
axes[1].set_ylabel("Probability Density f(Z)", fontsize=12)
axes[1].legend(fontsize=10, loc='upper left')
axes[1].grid(alpha=0.3)

# --- Plot 3: Analytical Summary ---
axes[2].axis('off')
summary_text = (
    f"Proportion of Pregnancies Lasting Between 240 and 270 Days\n"
    f"{'-'*52}\n"
    f"Given Parameters:\n"
    f"  • Mean (μ) = 266 days\n"
    f"  • Standard Deviation (σ) = 16 days\n\n"
    f"Standardized Values:\n"
    f"  • Z₁ = (240 - 266) / 16 = {z1:.2f}\n"
    f"  • Z₂ = (270 - 266) / 16 = {z2:.2f}\n\n"
    f"Cumulative Probabilities:\n"
    f"  • P(Z < {z2:.2f}) = {p2:.4f}\n"
    f"  • P(Z < {z1:.2f}) = {p1:.4f}\n\n"
    f"Result:\n"
    f"  • P(240 < X < 270) = {p_between:.4f}\n"
    f"  • ≈ {p_between*100:.2f}% of pregnancies\n\n"
    f"Interpretation:\n"
    f"  • Roughly 55% of pregnancies last between 240–270 days.\n"
    f"  • This corresponds to approximately 8–9 months.\n"
    f"  • The remaining 45% fall outside this range.\n\n"
    f"Insight:\n"
    f"  • Standardization (Z-scores) enables comparison across contexts.\n"
    f"  • Bounded probability estimation is fundamental in biostatistics.\n"
    f"  • Such models inform clinical forecasting, policy, and risk assessment."
)
axes[2].text(0.02, 0.95, summary_text, fontsize=11.5, fontfamily='monospace', va='top')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Example: Pregnancy Duration Corresponding to the Shortest 70%
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -----------------------------
# GIVEN PARAMETERS
# -----------------------------
mu = 266      # Mean pregnancy duration (days)
sigma = 16    # Standard deviation (days)
p = 0.70      # Proportion (70%)

# -----------------------------
# STEP 1: FIND CORRESPONDING Z-SCORE
# -----------------------------
z = norm.ppf(p)   # Inverse CDF (70th percentile)
x = mu + z * sigma  # Convert to actual duration

# -----------------------------
# STEP 2: PREPARE DATA FOR VISUALIZATION
# -----------------------------
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_vals = norm.pdf(x_vals, mu, sigma)

# Shaded region for shortest 70%
x_shade = x_vals[x_vals <= x]
pdf_shade = norm.pdf(x_shade, mu, sigma)

# -----------------------------
# STEP 3: VISUALIZATION
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Normal Distribution Analysis: Shortest 70% of Pregnancies", fontsize=16, fontweight='bold')

# --- Plot 1: Normal Distribution (X-scale) ---
axes[0].plot(x_vals, pdf_vals, color='navy', lw=2, label="Normal PDF (μ=266, σ=16)")
axes[0].fill_between(x_shade, pdf_shade, color='lightgreen', alpha=0.7, label="Shortest 70% (P(X ≤ 274.3))")
axes[0].axvline(mu, color='red', linestyle='--', lw=1.5, label=f"Mean μ = {mu}")
axes[0].axvline(x, color='black', linestyle=':', lw=2, label=f"70th Percentile x = {x:.2f}")
axes[0].set_title("Pregnancy Duration Distribution", fontsize=13)
axes[0].set_xlabel("Pregnancy Duration (days)", fontsize=12)
axes[0].set_ylabel("Probability Density f(X)", fontsize=12)
axes[0].legend(fontsize=10, loc='upper left')
axes[0].grid(alpha=0.3)

# --- Plot 2: Analytical Summary ---
axes[1].axis('off')
summary = (
    f"Statistical Summary\n"
    f"{'-'*30}\n"
    f"Distribution: X ~ N(μ=266, σ=16)\n\n"
    f"Percentile Definition:\n"
    f"  P(X ≤ x) = 0.70\n"
    f"  ⇒ P(Z ≤ z) = 0.70\n\n"
    f"From Z-table or inverse CDF:\n"
    f"  z = {z:.2f}\n\n"
    f"Transformation:\n"
    f"  x = μ + zσ = 266 + ({z:.2f})×16 = {x:.2f}\n\n"
    f"Result:\n"
    f"  x₀.₇₀ = {x:.2f} days\n"
    f"  → 70% of pregnancies last ≤ 274.3 days.\n\n"
    f"Interpretation:\n"
    f"  • The 70th percentile lies 0.52 SD above the mean.\n"
    f"  • The remaining 30% exceed 274.3 days.\n"
    f"  • This threshold defines a 'shorter' pregnancy duration.\n\n"
    f"Insight:\n"
    f"  • Percentile–Z mapping underpins quantile-based inference.\n"
    f"  • Enables estimation of medical or biological thresholds.\n"
    f"  • Useful for gestational risk analysis, policy setting,\n"
    f"    and population-based health forecasting."
)
axes[1].text(0.02, 0.95, summary, fontsize=11.5, fontfamily='monospace', va='top')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

##  **1. Types of Distributions Covered**

---

### **Discrete Uniform Distribution**

- A **random variable** that takes a finite number of **equally likely values**.  
- **Probability mass function (PMF):**

$$ P(X = x) = \frac{1}{k}, \quad \text{for } k \text{ outcomes} $$

**Example:** Tossing a fair die (each outcome 1–6 has equal probability).

---

### **Continuous Uniform Distribution**

- A **continuous random variable** can take any value in the interval \( [a, b] \).  
- **Probability density function (PDF):**

$$ f(x) = \frac{1}{b - a}, \quad a \le x \le b $$

- **Expected value:**

$$ E(X) = \frac{a + b}{2} $$

- **Variance:**

$$ \sigma^2 = \frac{(b - a)^2}{12} $$

**Example:** The amount of salad taken at a buffet (continuous range).

---

### **Binomial Distribution**

- Models the **number of successes** in \( n \) independent **Bernoulli trials**.  
- **PMF:**

$$ P(X = x) = \binom{n}{x} p^x (1 - p)^{n - x}, \quad x = 0, 1, \dots, n $$

**Example:** Number of defective items in a random sample.

---

### **Normal Distribution**

- Continuous, **bell-shaped**, and **symmetric** about the mean \( \mu \).  
- **Standard form:**

$$ X \sim N(\mu, \sigma^2) $$

- **Empirical Rule:**
  - 68% of data within \( \pm 1\sigma \)
  - 95% within \( \pm 2\sigma \)
  - 99.7% within \( \pm 3\sigma \)

- **Standardization:**

$$ Z = \frac{X - \mu}{\sigma} $$

---

### **Standard Normal Distribution**

- Special case of the normal distribution:

$$ Z \sim N(0, 1) $$

- Probabilities derived from **Z-tables**:
  - Provide cumulative areas under the standard normal curve.
- Enables **universal computation** of probabilities for any normal variable through standardization.

---

##  **2. Key Concepts & Interpretation**

| Concept | Discrete | Continuous |
|----------|-----------|------------|
| Outcomes | Finite / Countable | Infinite (within an interval) |
| Probability | Sum of probabilities | Area under the PDF |
| Example | Toss of a die | Height or weight |
| Representation | PMF | PDF |

- **Standardization:** Converts any normal variable to a **Z-score** for lookup in standard tables.  
- **Tail probabilities:** Used in risk analysis, hypothesis testing, and quality control.  
- **Interval probabilities:** Represent probabilities between two bounds, e.g., $$P(a \le X \le b) $$.
